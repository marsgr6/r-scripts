- El gradiente de una función escalar multivariable $f(x, y, \dots)$, denotado como $\nabla f$, que empaqueta toda la información de sus derivadas parciales en un vector:

![](https://cdn.kastatic.org/ka-perseus-images/d470e6f26c60f11da67fb71355c95aea642835c2.png)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfx = lambda x: 2*x
f = lambda x: x**2

# Define tangent line
# y = m*(x - x1) + y1
def line(x, x1, y1):
    return dfx(x1)*(x - x1) + y1

x = np.linspace(-4, 4, 100)

plt.figure(figsize=(8,8))
plt.plot(x, f(x))

lr = 0.2

x_0 = 4

iterations = 30

#print(x_0, f(x_0))

x_0_l = [x_0]
fx_0_l = [f(x_0)]

xrange = np.linspace(x_0-1, x_0+1, 10)
plt.scatter(x_0, f(x_0), color='C2', s=50)
plt.plot(xrange, line(xrange, x_0, f(x_0)), 'C2--', linewidth = 2)

for it in range(iterations):
    x_0 = x_0 - lr*(dfx(x_0))
    x_0_l += [x_0]
    fx_0_l += [f(x_0)]
    xrange = np.linspace(x_0-1, x_0+1, 10)
    plt.scatter(x_0, f(x_0), color='C2', s=50)
    plt.plot(xrange, line(xrange, x_0, f(x_0)), 'C2--', linewidth = 2)
    
plt.plot(x_0_l, fx_0_l, '-o')
plt.xlim(0, 5)
plt.ylim(0, 17)
#x_0_l, fx_0_l
```

```python
plt.semilogx(range(1, len(fx_0_l)+1), fx_0_l)
```
