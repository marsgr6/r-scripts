```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm
```

- Simulate pollen movements as a 2D random walk in python whose two components are independently normally distributed with expectation 0 mm and standard deviation 0.05 mm.

```python
mu = 0
sigma = 0.05

steps = 10
walks = 2

def make_walk(steps):
    x = np.concatenate(([0], norm.rvs(mu, sigma, steps))).cumsum()
    y = np.concatenate(([0], norm.rvs(mu, sigma, steps))).cumsum()
    return x, y

walks = []
for i in range(10000):
    walk = make_walk(100)
    walks += [walk]
    plt.plot(walk[0], walk[1])
```

```python
walks = np.array(walks)
np.shape(walks)
```

```python
distances = []
for walk in walks:
    distance = np.sqrt((walk**2).sum(axis=0))
    distances += [distance]
    plt.plot(distance)
distances = np.array(distances)
```

```python
plt.plot(distances.mean(axis=0))
```
