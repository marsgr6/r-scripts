```python
import numpy as np
import matplotlib.pyplot as plt
```

$f(x) = \frac{x^2}{3}, \quad -1 < x < 2$

$\int_{-\infty}^\infty f(x) dx = 1$

$\int_{-1}^2 \frac{x^2}{3} dx = 1$

$\int_{-1}^2 \frac{x^2}{3} dx = \frac{x^3}{9} |_{-1}^2 = \frac{(2)^3}{9} - \frac{(-1)^3}{9} = 
\frac{8}{9} - \frac{-1}{9} = \frac{9}{9} = 1$

$F(x) = P(X\leq x) = \int_{-\infty}^x f(t) dt$

$F(x) = \int_{-1}^x \frac{t^2}{3} dt = \frac{t^3}{9} |_{-1}^x = \frac{x^3}{9} + \frac{1}{9} = \frac{x^3+1}{9}$

$u = \frac{x^3+1}{9} = (9u - 1)^{1/3}$

$F^{-1}(u) = (9u - 1)^{1/3}$

```python
f = lambda x: x**2 / 3

F = lambda x: (x**3 + 1) / 9

ICDF = lambda u: np.cbrt(9*u-1)

x = np.linspace(-1, 2, 1000)

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(x, f(x))

a, b = 0, 1
x = np.linspace(a, b, 1000)

plt.fill_between(x, f(x), color="orange")

print("P({}<X<{})={}".format(a, b, np.trapz(f(x), x)))

plt.subplot(1,2,2)

x = np.linspace(-1, 2, 1000)

plt.plot(x, F(x))
plt.vlines([a, b], [0, 0], [F(a), F(b)], colors='k', linestyles='dotted')
plt.hlines([F(a), F(b)], [a, b], [-1, -1], colors='k', linestyles='dotted')

print("P({}<X<{})=F({})-F({})={}".format(a, b, b, a, F(b)-F(a)))
```

```python
u = np.random.random(size=100000)
h = plt.hist(ICDF(u), bins=40, edgecolor="orange")
```

```python
a, b, c = -1, 2, 2
X = []
n = 10000
while len(X) < n:
    xr = np.random.random()*(b-a) + a
    yr = np.random.random()*c
    if yr < f(xr):
        X += [xr]
h = plt.hist(X, bins=40, edgecolor="orange")
```
