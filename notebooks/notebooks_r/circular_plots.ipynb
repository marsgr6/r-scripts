```r
# Gráficos circulares
# http://zuguang.de/circlize/
# install.packages("circlize")
library(circlize)

# Define colores para los sectores
grid.col = c("#eb4827", "#804c79", "#6c8dc5", "#682de7", "#d4e30a", "#78b453")

# Descargar datos:
# https://drive.google.com/open?id=0B-dmEvN59Z2ZLUZ5Um1odld1cWM
# Lee archivo como una matriz, esta es la entrada que 
# se requiere una matriz con nombres en columnas y filas
mat=as.matrix(read.table("circleASIAtopSix.txt"))

# Define los nombres de la matriz para filas y columnas
# estos se usaran para los nombres de los sectores
colnames(mat) = c("China", "Japan", "South Korea", "India", "Singapore", "Thailand")
rownames(mat) = c("China", "Japan", "South Korea", "India", "Singapore", "Thailand")

chordDiagram(mat, grid.col = grid.col, annotationTrack = "grid", 
             preAllocateTracks = list(track.height = 0.1), transparency = 0.3, directional = TRUE)

circos.trackPlotRegion(track.index = 1, panel.fun = function(x, y) {
    xlim = get.cell.meta.data("xlim")
    xplot = get.cell.meta.data("xplot")
    ylim = get.cell.meta.data("ylim")
    sector.name = get.cell.meta.data("sector.index")
    if(abs(xplot[2] - xplot[1]) > 40) {
        circos.lines(xlim, c(mean(ylim), mean(ylim)), lty = 3) # dotted line
        for(p in seq(0.1, 0.9, by = 0.1)) {
            circos.text(p*(xlim[2] - xlim[1]) + xlim[1], mean(ylim) + 0.1,
            p, cex = 0.5, adj = c(0.5, 0), niceFacing = TRUE)
        }
    }
    circos.text(mean(xlim), 1, sector.name, niceFacing = TRUE, adj = c(0.5, 0))
}, bg.border = NA)

circos.trackPlotRegion(track.index = 1, panel.fun = function(x, y) {
    xlim = get.cell.meta.data("xlim")
    xplot = get.cell.meta.data("xplot")
    ylim = get.cell.meta.data("ylim")
    sector.name = get.cell.meta.data("sector.index")
    if(abs(xplot[2] - xplot[1]) > 20 & abs(xplot[2] - xplot[1]) < 40) {
        circos.lines(xlim, c(mean(ylim), mean(ylim)), lty = 3) # dotted line
        for(p in seq(0.1, 0.9, by = 0.2)) {
            circos.text(p*(xlim[2] - xlim[1]) + xlim[1], mean(ylim) + 0.1,
            p, cex = 0.5, adj = c(0.5, 0), niceFacing = TRUE)
        }
    }
    circos.text(mean(xlim), 1, sector.name, niceFacing = TRUE, adj = c(0.5, 0))
}, bg.border = NA)


circos.trackPlotRegion(track.index = 1, panel.fun = function(x, y) {
    xlim = get.cell.meta.data("xlim")
    xplot = get.cell.meta.data("xplot")
    ylim = get.cell.meta.data("ylim")
    sector.name = get.cell.meta.data("sector.index")
    if(abs(xplot[2] - xplot[1]) > 5 & abs(xplot[2] - xplot[1]) < 20) {
        circos.lines(xlim, c(mean(ylim), mean(ylim)), lty = 3) # dotted line
        for(p in seq(0, 1, by = 0.5)) {
            circos.text(p*(xlim[2] - xlim[1]) + xlim[1], mean(ylim) + 0.1,
            p, cex = 0.5, adj = c(0.5, 0), niceFacing = TRUE)
        }
    }
    circos.text(mean(xlim), 1, sector.name, niceFacing = TRUE, adj = c(0.5, 0))
}, bg.border = NA)

circos.trackPlotRegion(track.index = 1, panel.fun = function(x, y) {
    xlim = get.cell.meta.data("xlim")
    xplot = get.cell.meta.data("xplot")
    ylim = get.cell.meta.data("ylim")
    sector.name = get.cell.meta.data("sector.index")
    if(abs(xplot[2] - xplot[1]) < 5) {
        circos.lines(xlim, c(mean(ylim), mean(ylim)), lty = 3) # dotted line
        for(p in seq(0, 1, by = 1)) {
            circos.text(p*(xlim[2] - xlim[1]) + xlim[1], mean(ylim) + 0.1,
            p, cex = 0.5, adj = c(0.5, 0), niceFacing = TRUE)
        }
    }
    circos.text(mean(xlim), 1, sector.name, niceFacing = TRUE, adj = c(0.5, 0))
}, bg.border = NA)

circos.clear()

# dev.copy2pdf(file = "ASIAtopSix.pdf")
```
