```r
# Código R para graficar PMF y CDF
# Linea que inicia con # es un comentario
# Instalar R o usar el siguiente enlace
# https://www.tutorialspoint.com/execute_r_online.php

x = c(0,1,2,3,4,5,6) # rango de la v.a., x es un vector
px = c(0.10,0.15,0.20,0.25,0.20,0.06,0.04) # p(x): función de masa de probabiliad

# Plot PMF x vs px
plot(x,px,type="h",xlab="x",ylab="PMF",ylim=c(0,0.26)) 
# type="h" para histograma
# xlab y ylab para las etiquetas de los ejes
# ylim definie un límite inferior y superior para el eje y 
#   c(0,0.26) concatena los límites para ylim

points(x,px,ylim=c(0,0.26), pch=16) # agrega puntos al histograma

abline(h=0) # línea en el eje x (horizontal) para y=0

# Plot CDF
cdf = c(0,cumsum(px)) # suma acumulada, empezando en 0

cdf.plot = stepfun(x,cdf,f=0) # función escalón para la CDF discreta

# Grafica la función escalon generada en cdf.plot
plot.stepfun(cdf.plot,xlab="x",ylab="CDF",verticals=FALSE,do.points=TRUE,main="",pch=16)
```

```r
# Distribución binomial
# X ~ Bin(size=10, prob=0.5)

x = 0:10 # rango de la v.a., x es un vector, puede también ser un valor único
px = dbinom(x, size=10, prob=0.5) # p(x): función de masa de probabiliad
# dbinom:
#         la pmf será evaluada para el valor o valores en x, x es un vector
#         size: numero de experimentos
#         prob: probabilidad de éxito

# Plot PMF x vs px
plot(x,px,type="h",xlab="x",ylab="PMF",ylim=c(0,0.26)) 
# type="h" para histograma
# xlab y ylab para las etiquetas de los ejes
# ylim definie un límite inferior y superior para el eje y 
#   c(0,0.26) concatena los límites para ylim

points(x,px,ylim=c(0,0.26), pch=16) # agrega puntos al histograma

abline(h=0) # línea en el eje x (horizontal) para y=0

# Plot CDF
cdf = c(0,cumsum(px)) # suma acumulada, empezando en 0

cdf.plot = stepfun(x,cdf,f=0) # función escalón para la CDF discreta

# Grafica la función escalon generada en cdf.plot
plot.stepfun(cdf.plot,xlab="x",ylab="CDF",verticals=FALSE,do.points=TRUE,main="",pch=16)
```

```r
# Distribución geométrica
# X ~ Geom(prob=0.5)

x = 0:10 # rango de la v.a., x es un vector, puede también ser un valor único
px = dgeom(x, prob=0.5) # p(x): función de masa de probabiliad
# dbinom:
#         la pmf será evaluada para el valor o valores en x, x es un vector
#         size: numero de experimentos
#         prob: probabilidad de éxito

# Plot PMF x vs px
plot(x,px,type="h",xlab="x",ylab="PMF",ylim=c(0,0.525)) 
# type="h" para histograma
# xlab y ylab para las etiquetas de los ejes
# ylim definie un límite inferior y superior para el eje y 
#   c(0,0.26) concatena los límites para ylim

points(x,px,ylim=c(0,0.525), pch=16) # agrega puntos al histograma

abline(h=0) # línea en el eje x (horizontal) para y=0

# Plot CDF
cdf = c(0,pgeom(x, prob=0.5)) # suma acumulada, empezando en 0

cdf.plot = stepfun(x,cdf,f=0) # función escalón para la CDF discreta

# Grafica la función escalon generada en cdf.plot
plot.stepfun(cdf.plot,xlab="x",ylab="CDF",verticals=FALSE,do.points=TRUE,main="",pch=16)
```

```r
# Solución de ejemplos:

#Dist. Binomial
#La probabilidad de que un paciente se recupere de una rara
#enfermedad sanguínea es de 0.4. Si se sabe que 15 personas
#contrajeron la enfermedad, ¿cuál es la probabilidad de que 
# a) sobrevivan al menos 10, 
# b) sobrevivan de 3 a 8, y 
# c) sobrevivan exactamente 5?

# a) P(X>=10) = 1 - P(X<=9)
# usando la CDF pbinom, size: numero de experimentos, prob: probabilidad de éxito
print("Inciso a) sobrevivan al menos 10")
print("Usando la CDF")
1 - pbinom(9, size=15, prob=0.4)

print("Usando la PMF")
sum(dbinom(10:15, size=15, prob=0.4))
# Evaluamos para 

print("Inciso b) sobrevivan de 3 a 8")
print("Usando la CDF") # F(8) - F(2)
pbinom(8, size=15, prob=0.4) - pbinom(2, size=15, prob=0.4)

print("Usando la PMF")
sum(dbinom(3:8, size=15, prob=0.4))

print("Inciso c) sobrevivan exactamente 5")
print("Usando la CDF") # F(5) - F(4)
pbinom(5, size=15, prob=0.4) - pbinom(4, size=15, prob=0.4)

print("Usando la PMF")
dbinom(5, size=15, prob=0.4)

print("Inciso d) ¿Cuál es el máximo número de sobrevivientes con un 80% de certeza?")
# Quiero saber cual es el valor de x que acumula un valor >= 0.8
# P(X<=k) >= 0.8, quiero saber el valor que toma la variable aleatoria
# para la condición arriba
# en decir encontrar el valor de k para el cual se acumula
# un valor >= 0.8
print("Usando la Inversa de la CDF")
qbinom(0.8, size=15, prob=0.4) # Usando la inversa de la CDF qbinom

# Interpretemos el resultado usando la CDF
print("Probabilidad de que sobrevivan hasta 7 pacientes")
pbinom(7, size=15, prob=0.4) # Prob de que sobrevivan hasta 7 aun no acumula hasta el 0.8 de prob

print("Probabilidad de que sobrevivan hasta 8 pacientes")
pbinom(8, size=15, prob=0.4) # Prob de que sobrevivan 8 cumple el criterio P(X<=k) >= 0.8
```

```r
# Ejemplo dist. de Poisson

# El numero promedio de camiones-tanque que llega cada día a cierta
# ciudad portuaria es 10. Las instalaciones en el puerto pueden alojar a
# lo sumo 15 camiones-tanque por día. ¿Cuál es la probabilidad de que
# en un día determinado lleguen mas de 15 camiones y se tenga que
# rechazar algunos?

# X ~ Poisson(λ = 10)
# Se quiere P(X>15) = 1 - P(X<=15)

print("Usando la CDF") # 1 - F(15)
1 - ppois(15, lambda = 10)

print("Usando la PMF") 
1 - sum(dpois(0:15, lambda = 10))
```
