```r
# Distribucin gamma

x <- seq(0, 10, by=0.01)

pdfx <- dgamma(x, shape=1, scale = 2)
plot(x, pdfx, type='l')

pdfx <- dgamma(x, shape=2, scale = 2)
lines(x, pdfx, col="red")

pdfx <- dgamma(x, shape=3, scale = 2)
lines(x, pdfx, col="blue")

pdfx <- dgamma(x, shape=5, scale = 1)
lines(x, pdfx, col="green")

pdfx <- dgamma(x, shape=9, scale = 0.5)
lines(x, pdfx, col="purple")

legend('topright', c("k=1, theta=2.0", "k=2, theta=2.0", 
                     "k=3, theta=2.0", "k=5, theta=1.0", "k=9, theta=0.5"),
       lty=1, col=c('black', 'red', 'blue', 'green','purple'))
```

```r
# Distribucin t-student para diferentes grados de libertad

x <- seq(-5, 5, by=0.01)

pdfn <- dnorm(x)
plot(x, pdfn, type='l', col="black")

pdfx <- dt(x, df=1)
lines(x, pdfx, col="yellow")

pdfx <- dt(x, df=2)
lines(x, pdfx, col="green")

pdfx <- dt(x, df=5)
lines(x, pdfx, col="blue")

pdfx <- dt(x, df=10)
lines(x, pdfx, col="purple")

pdfx <- dt(x, df=30)
lines(x, pdfx, col="red")

legend('topright', c("N(0,1)", "T,v=1", 
                     "T,v=2", "T,v=2", "T,v=5", "T,v=10", "T,v=30"),
       lty=1, col=c('black', 'yellow', 'green', 'blue','purple', 'red'))
```
