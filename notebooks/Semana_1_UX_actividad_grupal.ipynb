### Actividad Sincrónica de Discusión - Semana 1

**Objetivo:**
El objetivo de esta actividad es promover la comprensión y discusión colaborativa sobre conceptos fundamentales en estadística descriptiva, gráficos descriptivos, análisis exploratorio de datos y visualización avanzada en el contexto de la analítica de datos.

**Instrucciones:**
1. **Formación de Grupos:**
   - Los participantes deben formar grupos de discusión en la plataforma Teams. Cada grupo deberá tener al menos 3 integrantes.
   - Cada grupo debe asignar a un moderador, quien será responsable de dirigir la discusión y mantener el orden.

2. **Discusión de Estadísticos Descriptivos:**
   - Discutan qué es un estadístico y su importancia en la analítica de datos.
   - Explore los estadísticos de centralidad y describan cuáles son y qué información proporcionan.
   - Analicen los estadísticos de dispersión y expliquen qué describen.
   - Aborden los estadísticos de orden y su utilidad en la descripción de datos.

3. **Discusión de Gráficos Descriptivos:**
   - Examinen los gráficos univariados y discutan cuáles pueden utilizarse para describir datos categóricos y continuos.
   - Analicen los gráficos multivariados y profundicen en los tipos de variables que pueden incorporarse.
   - Identifiquen y expliquen nombres específicos de gráficos multivariados, así como su interpretación.

4. **Análisis Exploratorio de Datos:**
   - Definan en qué consiste un análisis exploratorio de datos y por qué es crucial en el proceso de analítica de datos.
   - Expliquen las herramientas utilizadas para llevar a cabo un análisis exploratorio y cómo estas contribuyen al entendimiento de los datos.

5. **Visualización Avanzada:**
   - Introduzcan el concepto de visualización avanzada y discutan su importancia en la representación de datos complejos.
   - Analicen qué tipos de gráficos y técnicas de visualización avanzada pueden ser utilizados en situaciones específicas.
   - Compartan ejemplos de casos en los que la visualización avanzada ha proporcionado insights significativos.

6. **Intercambio de Conocimientos:**
   - Cada grupo deberá compartir sus conclusiones y aprendizajes con el resto de la clase a través de un resumen breve.
   - Animen a plantear preguntas y generar debate entre los grupos para enriquecer la comprensión de todos.

7. **Síntesis y Conclusiones:**
   - En la sesión plenaria, se discutirán las principales conclusiones de cada grupo.
   - Se destacarán las similitudes y diferencias en las interpretaciones y enfoques de los conceptos discutidos.

**Dinámica:**
La participación activa, la claridad en las explicaciones y la capacidad de síntesis serán bienvenidas. Se fomenta el debate y la colaboración entre los participantes para enriquecer la comprensión de los conceptos presentados, con especial énfasis en la visualización avanzada.
