## Data tidying

```
"Tidy datasets are all alike, but every messy dataset is messy in its own way." –– Hadley Wickham
```

El "data tidying" o limpieza de datos es un proceso fundamental en el ámbito de data science y data analytics. El propósito principal es preparar y organizar los datos de manera que sean adecuados para el análisis. Algunos de los objetivos específicos del data tidying incluyen:

1. **Facilitar el Análisis:** Datos bien organizados facilitan la exploración y análisis, permitiendo a los científicos de datos centrarse en la interpretación y obtención de información valiosa.

2. **Eliminar Inconsistencias:** Durante el proceso de recolección de datos, es común encontrar valores faltantes, errores tipográficos, o datos inconsistentes. La limpieza de datos busca corregir estos problemas para evitar interpretaciones erróneas.

3. **Consolidar Datos:** A veces, los datos pueden estar distribuidos en diferentes fuentes o formatos. La limpieza implica consolidar la información de manera coherente para que sea más fácil de analizar.

4. **Manejar Datos Faltantes:** Identificar y tratar adecuadamente los valores faltantes es crucial. Esto puede implicar la imputación de valores, la eliminación de observaciones o el diseño de estrategias específicas según el caso.

5. **Normalizar Formatos:** Asegurarse de que los datos estén en el formato correcto y que las unidades sean coherentes para que los resultados del análisis sean significativos.

6. **Crear Conjuntos de Datos Consistentes:** Los datos tidied se presentan en estructuras coherentes y fáciles de entender, lo que facilita su utilización por parte de otros profesionales o sistemas.

7. **Mejorar la Reproducibilidad:** La limpieza de datos bien documentada y reproducible es esencial para que otros científicos de datos o investigadores puedan comprender y replicar los análisis.

### Características de datos tydy

Los datos ordenados, también conocidos como datos "tidy" en el contexto de la limpieza y organización de datos, poseen ciertas características clave. Estas características facilitan el análisis y la interpretación de los datos. Las principales características de los datos ordenados son:

1. **Cada variable tiene su propia columna:** Cada característica o variable está representada en una columna separada. Esto facilita la identificación y extracción de información específica.

2. **Cada observación tiene su propia fila:** Cada instancia u observación individual se registra en una fila única. Esto ayuda a mantener la coherencia y facilita la comparación entre diferentes observaciones.

3. **Cada valor ocupa su propia celda:** Cada entrada en la tabla de datos se encuentra en su propia celda única. Esto asegura la organización y facilita el acceso a datos específicos.

4. **La tabla tiene una estructura rectangular:** Los datos están organizados en una estructura de tabla rectangular, donde todas las filas tienen el mismo conjunto de columnas. Esto simplifica el proceso de análisis y manipulación.

5. **La primera fila contiene encabezados:** La primera fila de la tabla generalmente contiene nombres de variables o encabezados, lo que proporciona información sobre el contenido de cada columna.

7. **Debe permitir operaciones eficientes:** La organización adecuada de los datos facilita la realización de operaciones como filtrado, agrupación, y análisis estadísticos, lo que contribuye a un flujo de trabajo más eficiente.

Estas características son fundamentales para garantizar que los datos sean manejables y comprensibles, lo que a su vez facilita la toma de decisiones informadas y la generación de conocimiento a partir de los datos.

## Operaciones comunes que se usan en data tydying

### Pivoteo en Data Tidying

El pivoteo es un proceso clave en la limpieza y organización de datos. Consiste en transformar un conjunto de datos para que sea más adecuado para el análisis, reorganizando las columnas y filas.

**Pivoteo Wide to Long:**

El pivoteo de "wide to long" es un proceso de reorganización de datos que transforma un conjunto de datos desde un formato ancho a un formato largo.

**Características del Formato Ancho (Wide):**
- Cada variable tiene su propia columna.
- Puede haber múltiples columnas que representan mediciones o condiciones en diferentes momentos.

**Pasos Comunes en el Pivoteo Wide to Long:**
1. **Identificar Variables Clave:** Seleccionar las variables que representan mediciones o condiciones en diferentes momentos.

2. **Derretir (Melt):** Combina estas variables en una sola columna y crea una columna adicional para almacenar los valores correspondientes.

3. **Crear una Columna de Tiempo o Condición:** Introducir una nueva variable que indique el momento o la condición asociada con cada valor.

**Ejemplo:**

Antes del pivoteo (wide):

| ID | T1_Score | T2_Score | T3_Score |
|----|----------|----------|----------|
| 1  | 80       | 85       | 90       |
| 2  | 75       | 78       | 80       |

Después del pivoteo (long):

| ID | Time | Score |
|----|------|-------|
| 1  | T1   | 80    |
| 1  | T2   | 85    |
| 1  | T3   | 90    |
| 2  | T1   | 75    |
| 2  | T2   | 78    |
| 2  | T3   | 80    |

En este ejemplo, hemos pivoteado los datos desde un formato ancho a uno largo, lo que facilita el análisis de las puntuaciones a lo largo del tiempo para cada individuo.

**Pivoteo Long to Wide:**

El pivoteo de "long to wide" es un proceso de reorganización de datos que transforma un conjunto de datos desde un formato largo a un formato ancho.

**Características del Formato Largo (Long):**
- Cada observación tiene su propia fila.
- Puede haber múltiples filas para la misma entidad, cada una representando una observación diferente.

**Pasos Comunes en el Pivoteo Long to Wide:**
1. **Identificar Variables Clave:** Seleccionar las variables que serán las nuevas columnas en el formato ancho.

2. **Pivotear:** Organizar los datos de manera que las observaciones únicas para cada entidad se conviertan en filas y las variables seleccionadas se conviertan en nuevas columnas.

3. **Lidiar con Valores Múltiples:** Decidir cómo manejar situaciones en las que hay múltiples observaciones para la misma entidad y variable (por ejemplo, agregación, toma del primer/último valor, etc.).

**Ejemplo:**

Antes del pivoteo (long):

| ID | Variable | Value |
|----|----------|-------|
| 1  | A        | 10    |
| 1  | B        | 15    |
| 2  | A        | 20    |
| 2  | B        | 25    |

Después del pivoteo (wide):

| ID | A   | B   |
|----|-----|-----|
| 1  | 10  | 15  |
| 2  | 20  | 25  |

En este ejemplo, hemos pivoteado los datos desde un formato largo a uno ancho, lo que facilita el análisis de las diferentes variables para cada entidad.

### Separating y Uniting en Data Tidying

**Separating:**

La operación de "separating" implica dividir una columna que contiene información combinada en dos o más columnas separadas. Esto es útil cuando una columna contiene valores que en realidad representan múltiples variables o categorías.

**Ejemplo:**
Antes de separar:

| ID | Name_Age     |
|----|--------------|
| 1  | John_25      |
| 2  | Alice_30     |

Después de separar:

| ID | Name   | Age |
|----|--------|-----|
| 1  | John   | 25  |
| 2  | Alice  | 30  |

### Preprocesamiento de Datos

El preprocesamiento de datos es un paso crucial en el análisis de datos y en el flujo de trabajo de aprendizaje automático. Involucra la limpieza y transformación de datos brutos en un formato adecuado para el análisis o para entrenar modelos de aprendizaje automático. El objetivo del preprocesamiento de datos es mejorar la calidad de los datos, aumentar su precisión y facilitar mejores resultados en pasos subsiguientes.

Tareas comunes en el preprocesamiento de datos incluyen:

1. **Manejo de Datos Faltantes:** Tratar los valores faltantes ya sea eliminándolos, imputándolos con valores estimados, o utilizando técnicas más avanzadas de imputación.

2. **Manejo de Valores Atípicos:** Identificar y manejar valores atípicos, que son puntos de datos que difieren significativamente del resto del conjunto de datos.

3. **Transformación de Datos:** Aplicar transformaciones a variables, como transformaciones logarítmicas o normalización, para hacer los datos más adecuados para el modelado.

4. **Codificación de Variables Categóricas:** Convertir variables categóricas a un formato numérico que pueda ser utilizado por algoritmos de aprendizaje automático. Esto puede incluir codificación one-hot, codificación de etiquetas u otros métodos.

5. **Escala y Estandarización:** Escalar características numéricas para asegurar que estén en una escala similar. Esto es especialmente importante para algoritmos sensibles a la magnitud de las características de entrada.

6. **Ingeniería de Características:** Crear nuevas características o modificar existentes para capturar información más relevante para el modelado.

7. **División de Datos:** Dividir el conjunto de datos en conjuntos de entrenamiento y prueba para evaluar modelos.

### Valores faltantes y cómo tratarlos

Los valores faltantes, también conocidos como valores nulos o valores perdidos, son aquellos que no están presentes en una variable o columna de datos en ciertas observaciones. Estos valores pueden surgir por diversas razones, como errores en la recopilación de datos, omisiones intencionales, o simplemente porque la información no está disponible.

**Tratamiento de Valores Faltantes:**

1. **Identificación de Valores Faltantes:**
   - Antes de abordar los valores faltantes, es importante identificar dónde se encuentran en el conjunto de datos. Puedes utilizar funciones o métodos específicos en tu plataforma de análisis de datos para detectarlos.

2. **Eliminación de Filas o Columnas:**
   - Si la cantidad de valores faltantes es pequeña en comparación con el tamaño total del conjunto de datos y no afecta significativamente el análisis, puedes optar por eliminar las filas o columnas que contienen valores faltantes.

3. **Imputación:**
   - La imputación implica estimar o predecir los valores faltantes utilizando información disponible. Puedes utilizar estadísticas descriptivas como la media, la mediana o la moda para llenar los valores faltantes. Otra opción es utilizar métodos más avanzados, como la imputación basada en modelos.

4. **Creación de Indicadores:**
   - En lugar de imputar valores faltantes, puedes agregar una columna adicional que actúe como indicador de la presencia de valores faltantes en la variable original. Esto permite conservar la información sobre la falta de datos.

5. **Manejo Contextual:**
   - En algunos casos, es posible que el valor faltante no sea simplemente una ausencia de datos, sino que tenga un significado específico. Por ejemplo, un campo de "Fecha de Finalización" no estaría lleno si la tarea aún no se ha completado. En estos casos, se puede optar por no tratar el valor faltante como un error, sino como una característica del conjunto de datos.

6. **Uso de Métodos Avanzados:**
   - En situaciones más complejas, especialmente en conjuntos de datos grandes y complejos, se pueden utilizar métodos más avanzados como la imputación múltiple o modelos de machine learning para tratar valores faltantes de manera más precisa.

La elección de la estrategia dependerá del contexto específico del conjunto de datos y del análisis que estás llevando a cabo. Es importante documentar y comunicar cómo se manejan los valores faltantes en tu análisis para garantizar la transparencia y la reproducibilidad.

**Técnicas de Imputación:**

Existen varias técnicas de imputación para tratar valores faltantes en conjuntos de datos. La elección de la técnica depende del tipo de datos y del contexto específico del problema. Aquí hay algunas técnicas comunes de imputación:

1. **Imputación por Media, Mediana o Moda:**
   - Esta técnica implica reemplazar los valores faltantes con la media, mediana o moda de la variable correspondiente. Es simple y adecuada para datos numéricos. Sin embargo, no tiene en cuenta la variabilidad entre las observaciones.

2. **Imputación por Vecinos más Cercanos (K-NN):**
   - Utiliza los valores de las observaciones más cercanas en términos de similitud para imputar los valores faltantes. Es eficaz tanto para variables numéricas como categóricas. Sin embargo, puede ser computacionalmente costoso y puede no funcionar bien en conjuntos de datos grandes o dispersos.

3. **Regresión Lineal o Múltiple:**
   - Se utiliza la regresión para predecir los valores faltantes en función de otras variables. Es efectiva cuando hay relaciones lineales en los datos, pero puede no funcionar bien si la relación no es lineal.

4. **Métodos de Imputación Múltiple (Multiple Imputation):**
   - Se generan varias imputaciones para los valores faltantes, y se realiza el análisis en cada conjunto imputado. Los resultados se combinan para obtener estimaciones más precisas. Es útil para manejar la incertidumbre asociada con los valores faltantes.

5. **Imputación Basada en Modelos (Model-based Imputation):**
   - Se utilizan modelos estadísticos o de machine learning para predecir los valores faltantes. Puede ser más preciso que la imputación por media o regresión, pero también puede ser más complejo y requerir más recursos computacionales.

6. **Imputación por Múltiples Métodos (MICE):**
   - Combina varias técnicas de imputación en un solo marco. En cada iteración, se imputan los valores faltantes utilizando un método específico, y este proceso se repite varias veces. Es flexible y puede adaptarse a la naturaleza de los datos.

7. **Imputación con Variables Dummy (Dummy Variable Imputation):**
   - Se crea una variable dummy que indica la presencia o ausencia de un valor faltante en la observación. Luego, se utiliza esta variable dummy junto con otras variables para imputar los valores faltantes.

8. **Imputación Hot-Deck:**
   - Consiste en emparejar observaciones con valores faltantes con observaciones similares que tengan valores conocidos. La información de las observaciones emparejadas se utiliza para imputar los valores faltantes.
   

**Técnicas de Imputación en Series Temporales:**

1. **Imputación Forward (Hacia Adelante):**
   - En la imputación forward, los valores faltantes se imputan utilizando información de observaciones previas en el tiempo. Es especialmente útil en series temporales donde la observación actual puede depender de valores pasados.

   *Ejemplo:*
   
|Tiempo	|Valor|
|----|--------------|
|T1|	10|
|T2	|NaN
|T3	|25|
|T4	|NaN|

En este caso, el valor faltante en T2 podría imputarse utilizando el valor en T1, ya que la serie temporal indica una cierta tendencia.

2. **Imputación Backward (Hacia Atrás):**
- En la imputación backward, los valores faltantes se imputan utilizando información de observaciones posteriores en el tiempo. Esta técnica puede ser útil si la observación actual se ve afectada por eventos futuros.

   *Ejemplo:*

|Tiempo|	Valor|
|----|--------------|
|T1|	10|
|T2|	NaN|
|T3|	25|
|T4|	NaN

En este caso, el valor faltante en T2 podría imputarse utilizando el valor en T3, ya que la serie temporal indica una cierta tendencia futura.

3. **Imputación Lineal:**
- Esta técnica utiliza la interpolación lineal para imputar los valores faltantes en una serie temporal. Se calcula una línea recta entre las observaciones adyacentes y se estima el valor faltante en función de esta línea.

4. **Media Móvil:**
- Se utiliza la media de las observaciones vecinas para imputar los valores faltantes. La ventana de la media móvil puede ser fija o variable, dependiendo de la naturaleza de la serie temporal.

5. **Imputación de Vecinos más Cercanos Temporales (Time-based k-NN):**
- Similar al enfoque de vecinos más cercanos en datos estáticos, esta técnica encuentra observaciones similares en términos temporales y utiliza sus valores para imputar los valores faltantes.

6. **Spline Interpolation:**
- Se ajusta una curva spline a la serie temporal para imputar los valores faltantes. Las splines son funciones suaves que se adaptan bien a patrones complejos en los datos temporales.

7. **Imputación Basada en Modelos (Model-based Imputation):**
- Utiliza modelos estadísticos o de machine learning para predecir y imputar los valores faltantes. Pueden ser modelos autoregresivos, modelos ARIMA u otros modelos más complejos.

8. **Imputación Estacional:**
- Considera patrones estacionales en la serie temporal y utiliza información estacional para imputar los valores faltantes.

9. **Imputación Múltiple:**
- Similar a la imputación múltiple en datos estáticos, se generan múltiples series temporales imputadas y se realizan análisis posteriores considerando la variabilidad entre las imputaciones.

10. **Métodos de Suavizado (Smoothing Methods):**
 - Emplea técnicas de suavizado, como el suavizado exponencial, para imputar valores faltantes. Estos métodos son especialmente útiles cuando hay patrones de tendencia y estacionalidad en la serie temporal.


La elección de la técnica de imputación depende de varios factores, como la naturaleza de los datos, la cantidad de valores faltantes y la complejidad del problema. Es recomendable evaluar el rendimiento de diferentes métodos en el contexto específico antes de seleccionar la técnica de imputación más adecuada.


### Discretización de Datos

La discretización de datos es el proceso de transformar variables continuas en variables discretas. Esto implica dividir el rango de valores de una variable continua en intervalos o categorías discretas. Aquí hay algunas técnicas comunes de discretización:

1. **Binning (Bandeo):**
   - Consiste en dividir el rango de valores en intervalos (bins) y asignar cada valor a uno de estos intervalos. Puede ser equidistante o basado en cuantiles.

2. **Equal-Width Binning (Bandeo de Ancho Equitativo):**
   - Divide el rango de valores en intervalos de ancho fijo. Por ejemplo, si tienes valores entre 0 y 100 y decides tener 5 bins, cada bin tendría un ancho de 20 unidades.

3. **Equal-Frequency Binning (Bandeo de Frecuencia Equitativa):**
   - Asigna aproximadamente el mismo número de observaciones a cada bin. Esto puede resultar en intervalos de ancho variable.

4. **K-Means Clustering:**
   - Utiliza el algoritmo de clustering K-Means para agrupar valores similares en k clusters y luego asigna cada valor al cluster correspondiente.

5. **Decision Trees:**
   - Puedes utilizar un árbol de decisión para identificar puntos de corte óptimos y discretizar la variable basándote en las divisiones del árbol.

6. **Custom Binning:**
   - Permite la creación de intervalos personalizados según el conocimiento del dominio o requisitos específicos del problema.

**Ejemplo:**
Supongamos que tenemos una variable continua "Edad" con valores entre 20 y 60, y queremos discretizarla en tres grupos: "Joven", "Adulto" y "Mayor". Podríamos aplicar una discretización personalizada para asignar valores a cada grupo según rangos específicos.

| Edad | Grupo   |
|------|---------|
| 25   | Joven   |
| 40   | Adulto  |
| 55   | Mayor   |
| ...  | ...     |

La elección de la técnica de discretización dependerá del contexto del problema y de los requisitos específicos del análisis o modelo.

**Discretización Ad Hoc Bajo Criterio de Experto**

La discretización ad hoc bajo criterio de experto es un enfoque en el cual se realiza la transformación de variables continuas en categorías discretas según el juicio y el conocimiento experto en el dominio específico. A diferencia de métodos automatizados, como binning equitativo o clustering, esta técnica depende completamente de la experiencia y la intuición de un experto en el campo.

**Proceso:**

1. **Evaluación del Contexto:**
   - El experto revisa el contexto del problema y comprende la naturaleza de la variable que se va a discretizar.

2. **Identificación de Puntos de Corte:**
   - El experto decide dónde establecer los puntos de corte para definir las categorías discretas. Esto puede basarse en conocimientos sobre el significado práctico de los valores, patrones observados o cualquier otro criterio relevante.

3. **Definición de Categorías:**
   - Se definen las categorías discretas y se asignan los valores a cada categoría según los puntos de corte establecidos. Esto puede implicar la creación de categorías específicas con significados relevantes en el dominio.

**Ejemplo:**
Supongamos que tenemos una variable de ingresos mensuales y queremos discretizarla. Un experto en economía podría decidir que las categorías serán "Bajo", "Medio" y "Alto" según sus conocimientos sobre los estándares económicos y las condiciones locales.

| Ingresos | Categoría |
|----------|-----------|
| 800      | Bajo      |
| 2000     | Medio     |
| 5000     | Alto      |
| ...      | ...       |

Este enfoque ad hoc es particularmente útil cuando la discretización no puede ser fácilmente definida por reglas automáticas y requiere un entendimiento profundo del dominio.

La discretización ad hoc bajo criterio de experto proporciona flexibilidad y permite adaptarse a situaciones específicas que pueden no ser capturadas por métodos automáticos.


### Escalado

La escala es un aspecto específico del preprocesamiento de datos que se centra en transformar características numéricas a una escala estándar. El objetivo es asegurar que ninguna característica en particular domine a las demás debido a diferencias en sus magnitudes. La escala es especialmente importante para algoritmos que dependen de medidas de distancia, optimización mediante descenso de gradiente o cualquier método sensible a la escala de las características de entrada.

Técnicas comunes de escala incluyen:

1. **Escala Min-Max (Normalización):** Escalar características a un rango específico, a menudo entre 0 y 1.

2. **Estandarización (Normalización Z-score):** Escalar características para que tengan una media de 0 y una desviación estándar de 1.

3. **Escala Robusta:** Escalar características basadas en percentiles, siendo menos sensible a valores atípicos.

La elección del método de escala adecuado depende de las características de los datos y de los requisitos del algoritmo de aprendizaje automático que se esté utilizando. En general, el preprocesamiento de datos, incluida la escala, es un paso crucial para preparar los datos para el análisis y el entrenamiento de modelos, asegurando que los datos estén en un formato adecuado para obtener resultados precisos y significativos.

## Ejemplo práctico: Preparación de un Conjunto de Datos para Modelos Predictivos de Calidad del Aire

El conjunto de datos que se analizará proviene de la Secretaría de Ambiente de la Municipalidad de Quito, Ecuador. Se incluyeron diversas variables meteorológicas y de química atmosférica en el análisis.

A continuación, se detallan las variables utilizadas en este ejercicio:

- **PM2.5:** 
La variable PM2.5 se refiere a las partículas suspendidas en el aire que tienen un diámetro aerodinámico de 2.5 micrómetros o menos. Estas partículas son consideradas material particulado fino y pueden incluir polvo, hollín, cenizas, y otros contaminantes atmosféricos. La concentración de PM2.5 en el aire se mide en microgramos por metro cúbico (µg/m³) y se utiliza como indicador de la calidad del aire. Altas concentraciones de PM2.5 están asociadas con problemas respiratorios, cardiovasculares y otros efectos adversos para la salud.

- **Velocidad de viento:** La velocidad del viento puede tener un impacto significativo en la concentración de partículas PM2.5 en el aire. La velocidad del viento se refiere a la magnitud de la velocidad con la que el aire se mueve en la atmósfera. Se mide comúnmente en unidades como metros por segundo **(m/s), que es nuestro caso**, kilómetros por hora (km/h) o millas por hora (mph). 

- **Dirección del Viento:** La dirección del viento se refiere a la dirección desde la cual el viento está soplando. Se expresa en grados, donde 0 grados generalmente indica que el viento proviene del norte y se mide en sentido contrario a las agujas del reloj. Algunos puntos de referencia comunes son:

  - **0 grados:** Norte
  - **90 grados:** Este
  - **180 grados:** Sur
  - **270 grados:** Oeste

  La dirección del viento es esencial para comprender cómo se desplazan los patrones atmosféricos, afectando la dispersión de contaminantes. 

- **Precipitación:** La cantidad de precipitación se mide comúnmente en milímetros o pulgadas, representando la altura de agua acumulada sobre una superficie plana durante un período específico. a precipitación, especialmente en forma de lluvia, actúa como un mecanismo natural de "lavado" del aire. Al caer, las gotas de lluvia capturan y arrastran consigo las partículas suspendidas, incluyendo las PM2.5, limpiando el aire.

Este proceso es útil para trabajar con datos en un formato más adecuado para análisis y visualización.

Brevemente, el `melt` realiza las siguientes operaciones:

Este proceso es particularmente útil cuando se trabaja con datos en formato ancho (muchas columnas) y se desea cambiar a un formato largo para facilitar ciertos tipos de análisis y visualizaciones.

### Unión de variables

En pandas, las operaciones de cadena (string) se refieren a las funciones y métodos que se pueden aplicar a las columnas de tipo cadena en un DataFrame. Estas operaciones permiten realizar manipulaciones, transformaciones y análisis de datos basados en texto. A continuación, se describen algunas de las operaciones de cadena comunes en pandas:

- **Concatenación de Cadenas:**
`str.cat()`: Concatena las cadenas de dos Series.
- **Extracción de Subcadenas:**
`str.slice()`, `str[índice_inicio:índice_fin]`: Extraen porciones específicas de las cadenas.
- **Mayúsculas y Minúsculas:**
`str.upper()`: Convierte a mayúsculas.
`str.lower()`: Convierte a minúsculas.
`str.capitalize()`: Convierte la primera letra a mayúscula.
- **Reemplazo y Eliminación:**
`str.replace()`: Reemplaza parte de una cadena por otra.
`str.strip()`: Elimina espacios en blanco al inicio y al final de la cadena.
- **Búsqueda y Conteo:**
`str.contains()`: Verifica si una cadena contiene otra.

- **Formato y Conversión:**
`str.format()`: Formatea una cadena con valores específicos.
`str.extract()`: Extrae patrones específicos utilizando expresiones regulares.

Vamos a utilizar estos métodos para unir las variables `['ANO', 'MES', 'DIA', 'HORA']`  en una sola columna `DATE_TIME`.

````
# Convertimos a string las columnas especificadas
pm25m[['ANO', 'MES', 'DIA']] = pm25m[['ANO', 'MES', 'DIA']].astype(str)
pm25m['MES'] = pm25m['MES'].str.rjust(2,'0')  # Agregamos padding de 0s
pm25m['DIA'] = pm25m['DIA'].str.rjust(2,'0')  # Agregamos padding de 0s

# Unimos DIA MES y ANO usando como separado /
pm25m['DATE'] = pm25m['DIA'].str.cat([pm25m['MES'], pm25m['ANO']], sep='/')

# Removemos HORA de la columna variable y agregamos padding de 0
# y sumamos minutos y segundos
pm25m['HORA'] = pm25m['variable'].str.replace('HORA', '')

# Hora va de 1 a 24, resto 1 para tener 0 a 23
# antes debo transformar el string a numérico para restar
# y vuelvo a trasnformar a string para concatenar mins y seg
pm25m['HORA'] =  (pd.to_numeric(pm25m['HORA'])-1).astype(str).str.rjust(2,'0') + ':00:00'

pm25m['DATE_TIME'] =  pm25m['DATE'].str.cat(pm25m['HORA'], sep=' ')

pm25m.head(3)
````

````
# Seleccionamos las columnas que nos interesan
datos_final = pm25m[['ESTACION', 'DATE_TIME', 'value']]

# Cambiamos nombres a las columnas
datos_final = datos_final.rename(columns={'ESTACION': 'Station', 
                            'DATE_TIME': 'Date_time', 'value': 'pm25ugm3'})

# Guardar archivo csv
datos_final.to_csv('./data/03pm25ugm3.csv', index=False)

datos_final.head(3)
````

### Merge dataframes

La función `merge` en pandas se utiliza para combinar dos DataFrames en función de una o más columnas comunes, similar a las operaciones de JOIN en SQL. Permite realizar operaciones de fusión horizontal, donde se unen las filas de dos DataFrames en función de valores coincidentes en una o más columnas. A continuación, se explican los parámetros clave y su uso:

```Python
result = pd.merge(left, right, how='inner', on=None, left_on=None, 
                  right_on=None, left_index=False, right_index=False, 
                  suffixes=('_left', '_right'), sort=False)
```

- `left` y `right`: Los DataFrames que se van a combinar.
`how`: Especifica el tipo de fusión y puede ser **'inner'** (intersección), **'outer'** (unión), **'left'** (izquierda) o **'right'** (derecha).
- `on`: Columna o lista de columnas en las que se basará la fusión.
- `left_on` y `right_on`: Columnas del DataFrame izquierdo y derecho, respectivamente, para fusionar si las columnas de fusión tienen nombres diferentes.
- `left_index` y `right_index`: Si son **True**, utiliza los índices del DataFrame izquierdo o derecho, respectivamente, como las columnas de fusión.
- `suffixes`: Sufijos a agregar a los nombres de columnas en caso de conflicto.
- `sort`: Si es **True**, ordena el resultado según las columnas de fusión.

- La ingeniería de características es un proceso en el que se crean nuevas variables (características) a partir de las existentes para mejorar el rendimiento de un modelo o facilitar su interpretación.

- En el contexto de la dirección y velocidad del viento, convertir las coordenadas polares a cartesianas es un ejemplo de ingeniería de características que puede hacer que los datos sean más accesibles para algunos algoritmos de aprendizaje automático.

- En meteorología, la dirección del viento y la velocidad del viento a menudo se representan en coordenadas polares, donde la dirección es el ángulo y la velocidad es la distancia desde el origen. Convertir estas coordenadas polares a cartesianas (coordenada X, Y) puede simplificar el manejo de la información y permitir que ciertos algoritmos aprovechen mejor los datos. 

- La variable a modelar es continua (PM2.5: `PARTICULAS(microg/m3))`, por lo tanto estamos frente a un problema de regresión. 
  - Podemos expresar esta variable de forma categórica usando umbralización a partir de valores dados por expertos.
  - Ej.: La OMS, la regulación del municipio, entre otros. 
  - En este caso estaríamos frente a un problema de clasificación.
  
- Esta discritización será realizada usando `np.where` para crear las columnas `PL2` y `PL3` para un problema de clasificación binaria (clases H: Healthy, UH: Unhealthy) y multiclase (3 clases: L: Low, M: Medium, H: High contamination).
  - En pandas, `np.where` se utiliza para realizar operaciones basadas en condiciones y asignar valores en función de esas condiciones. `np.where` es parte de la biblioteca NumPy, y se utiliza para realizar evaluaciones booleanas en un array NumPy y devolver valores basados en esas evaluaciones. 

````
# Polar to Cartesian 
datos['X'] = np.cos(datos['WINDDIR(deg)']*np.pi/180)*datos['WINDSPEED(m/s)']  # wind speed is the radius
datos['Y'] = np.sin(datos['WINDDIR(deg)']*np.pi/180)*datos['WINDSPEED(m/s)']
# WHO says that over 15 is bad (the expert).
datos['PL2'] = np.where(datos['PARTICULAS(microg/m3)']>=15, 'UH', 'H')
datos['PL3'] = np.where(datos['PARTICULAS(microg/m3)']<10, 'L', 
                        np.where(datos['PARTICULAS(microg/m3)']<25, 'M', 'H'))
datos.head(10)
````

````
# Guardamos ciertas columnas en dependencia del analisis
# que se realizará eventualmente
# Problema de regresión
datos_regresion = datos[['X', 'Y', 'PRECIPITATION(mm)', 'PARTICULAS(microg/m3)']]
# Problema de clasificación binaria
datos_binclass = datos[['X', 'Y', 'PRECIPITATION(mm)', 'PL2']]
# Problema multiclase
datos_mulclass = datos[['X', 'Y', 'PRECIPITATION(mm)', 'PL3']]
datos_regresion.to_csv("regressionData_pandas.csv", index=False)
datos_binclass.to_csv("binClassData_pandas.csv", index=False)
datos_mulclass.to_csv("mulClassData_pandas.csv", index=False)
````

Un scatterplot (gráfico de dispersión) con `hue` es una representación gráfica que utiliza puntos para visualizar la relación entre dos variables, y utiliza un tercer atributo (`hue`) para colorear los puntos según un valor específico. Esta técnica es especialmente útil cuando deseas explorar la relación entre dos variables mientras también consideras la contribución de una tercera variable categórica o continua.

- Variables `x` e `y`: Representan las variables que se comparan en los ejes horizontal y vertical.

- `hue`: Especifica la variable continua que se utiliza para asignar colores a los puntos.

- Mapa de colores (cmap): Determina la paleta de colores utilizada para representar los valores continuos.

- Barra de color (colorbar): Proporciona una referencia visual de cómo se asignan los colores a los valores continuos.

- Si la variable hue es categórica el gráfico colorea los puntos según la categoría. Cada categoría se representa con un color diferente en el gráfico.


- `xlim` y `ylim`: Límites de los ejes x e y.
- `fontsize`: Tamaño de la fuente.
- `colormap`: Colormap para tramas de líneas.

- Lectura de archivo con una columna de tipo de datos que se usará como índice temporal

  - **`./data/merged_data_no_missing_values.csv`**: Este es el camino relativo o absoluto al archivo CSV que estás leyendo. Asegúrate de que el archivo existe en esa ubicación.

  - **`index_col=[1]`**: Esto establece que la segunda columna del archivo CSV (índice 1 en base cero) se utilizará como el índice del DataFrame. Es decir, los valores en esa columna se usarán como etiquetas de fila en el DataFrame.

  - **`parse_dates=[1]`**: Este argumento indica a pandas que debe intentar interpretar las fechas en la segunda columna del archivo CSV como objetos de fecha y hora.   - **`dayfirst=True`**: Este argumento se utiliza en conjunto con parse_dates. Indica a pandas que, si las fechas tienen el formato día-mes-año, debe interpretarlas de esa manera. Es decir, si tienes una fecha como "01-02-2023", pandas la interpretará como el 1 de febrero de 2023.

La expresión `resample('M').sum()` se utiliza en pandas para realizar una operación de remuestreo temporal sobre un DataFrame con fechas y luego calcular la suma para cada mes ('M' significa mensual). 

- **`resample('M')`**: Este método se utiliza para cambiar la frecuencia temporal de los datos. En este caso, 'M' indica que queremos remuestrear a una frecuencia mensual. Si los datos tienen fechas diarias, este método agrupará los datos en meses.

- **`.sum()`**: Después de aplicar el remuestreo, .sum() se utiliza para calcular la suma de los valores en cada mes. Esto es especialmente útil si los datos representan cantidades acumulativas diarias y deseas obtener la suma mensual.

````
m_summary = pd.DataFrame()
m_summary['precipitation.mm'] = belisario['precipitation.mm'].resample('M').sum()
m_summary.head()
````

````
plt.plot(m_summary.index, m_summary['precipitation.mm'], '-')
plt.show()
````

````
import ipywidgets
import plotly.express as px
import plotly.graph_objects as go

@ipywidgets.interact
def plot(var_x=data.columns[1:],
         station=np.unique(data.Station),
         resample=['D', 'M', 'Q', 'Y'],
        ):
    df = data.query("Station=='station'").drop(columns="Station")
    df_r = df[var_x].resample(resample).sum()
    fig = go.FigureWidget()
    fig = px.line(df_r, x=df_r.index, y=df_r[var_x], title=var_x)
````
