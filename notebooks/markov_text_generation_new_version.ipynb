```python
table = str.maketrans('', '', string.punctuation)
stripped = [w.translate(table) for w in words]
words_min = [word.lower() for word in stripped]
print(words_min)
```

```python
import numpy as np
import pandas as pd
words_names = np.unique(words_min)
p_unicas = len(words_names)
matrix = np.zeros((p_unicas, p_unicas))
df = pd.DataFrame(matrix, columns=words_names,
                  index=words_names
                 )
for i in range(1, len(words_min)):
    col = words_min[i]
    row = words_min[i-1]
    df[col][row] += 1

df
```

```python
df.loc[:,:] = df.loc[:,:].div(df.sum(axis=1), axis=0)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.from_numpy_matrix(dfb.values, create_using=nx.DiGraph())

label_mapping = {idx: val for idx, val in enumerate(dfb.columns)}

G = nx.relabel_nodes(G, label_mapping)

plt.figure(figsize=(8,8))
nx.draw_kamada_kawai(G, with_labels=True, node_color='orange')
```

```python
def rw_graph(start, steps, states, df):
    walk = [start]
    for si in range(steps):
        p = df.loc[walk[-1], :]
        if np.isnan(p[0]):
            break
        else:
            walk += [np.random.choice(states, p=p)]
    return walk
```

```python
for i in range(10):
    w = ' '.join(
        rw_graph('only', 8, words_names, df))
    print(w)
```
