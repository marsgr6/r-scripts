### Grafos

- Un grafo es una colección de nodos, ó vértices (vertex) interconectados por ramas (edge):
  
  ![](https://upload.wikimedia.org/wikipedia/commons/2/2f/Small_Network.png)

- Uno de los modelos más importantes en teoría de grafos y redes complejas es el de un grafo aleatorio. 

  - Las grandes estructuras complejas que se encuentran en la naturaleza suelen estar formadas por interacciones individuales no correlacionadas entre un gran número de sus constituyentes, por lo que no sorprende que los grafos aleatorios proporcionen buenos modelos para muchas de ellas. 

  - Aunque las estructuras hechas por el hombre son el resultado de un diseño racional (o así lo deseamos), a menudo exhiben suficiente "aleatoriedad", ya sea porque su estructura se desarrolló con el tiempo (como la World-Wide-Web), o porque copian un modelo anterior de red existente (por ejemplo, red aeroportuaria, red eléctrica) o están sujetas a restricciones aleatorias o de otro tipo (por ejemplo, geográficas). 

  - Los grafos aleatorios con las características adecuadas pueden servir como modelos útiles para la evolución y el funcionamiento de muchas estructuras artificiales complejas.

### Modelado de datos (ajuste de distribuciones)

El ajuste de distribuciones, en el contexto de estadísticas y análisis de datos, se refiere al proceso de encontrar el modelo de distribución que mejor se ajusta a un conjunto de datos observado. En otras palabras, consiste en identificar la función de probabilidad que describe de manera óptima la variabilidad y la estructura de los datos recopilados.

Los pasos comunes en el ajuste de distribuciones incluyen:

1. **Selección del Modelo:** Se elige un tipo específico de distribución (por ejemplo, normal, exponencial, logística) que se presume podría describir la variabilidad de los datos.

2. **Estimación de Parámetros:** Se estiman los parámetros de la distribución seleccionada para que se ajusten mejor a los datos observados. Los parámetros pueden incluir la media, la desviación estándar, la tasa de éxito (en el caso de distribuciones binomiales), entre otros, dependiendo del tipo de distribución.

3. **Validación del Ajuste:** Se realizan pruebas estadísticas para evaluar cuán bien se ajusta la distribución propuesta a los datos reales. Esto podría incluir pruebas de bondad de ajuste, como la prueba de Kolmogorov-Smirnov o la prueba de chi-cuadrado.

4. **Iteración si es Necesario:** Si el ajuste inicial no es satisfactorio, se pueden probar diferentes distribuciones o se puede ajustar la estrategia de ajuste de parámetros para mejorar la calidad del ajuste.

El ajuste de distribuciones es fundamental en diversas áreas, como la inferencia estadística, la modelización de datos financieros, la simulación y en general, en cualquier situación en la que se busque caracterizar la variabilidad de los datos mediante una distribución probabilística.

**Objetivos del Ajuste de Distribuciones:**

1. **Descripción de Datos:** Proporcionar una descripción resumida y comprensible de un conjunto de datos mediante la identificación de la distribución de probabilidad que mejor se ajusta a ellos.

2. **Inferencia Estadística:** Facilitar la inferencia estadística al modelar los datos observados con distribuciones conocidas, lo que permite realizar predicciones y estimaciones sobre la población subyacente.

3. **Simulación:** Permitir la generación de datos sintéticos que sigan la misma distribución que los datos observados, lo que es útil para realizar simulaciones y evaluar escenarios hipotéticos.

4. **Análisis de Riesgos y Pérdidas:** En contextos financieros, el ajuste de distribuciones se utiliza para modelar la variabilidad de los rendimientos, pérdidas y otros factores relacionados con el riesgo, ayudando en la toma de decisiones y la gestión de riesgos.

5. **Detección de Anomalías:** Identificar valores atípicos o anomalías en los datos al comparar la distribución ajustada con las observaciones reales.

6. **Optimización de Procesos:** En ingeniería y manufactura, el ajuste de distribuciones puede ayudar a comprender y mejorar procesos mediante la modelización de variables críticas.

**Aplicaciones del Ajuste de Distribuciones:**

1. **Finanzas:** Modelar la variabilidad de los rendimientos de activos financieros para evaluar riesgos y rendimientos esperados.

2. **Calidad y Procesos Industriales:** Identificar patrones de variabilidad en variables de proceso para mejorar la calidad y eficiencia.

3. **Ciencias de la Salud:** Ajustar distribuciones a datos médicos para entender la distribución de características clínicas y epidemiológicas.

4. **Meteorología:** Modelar la distribución de eventos climáticos, como precipitaciones, temperaturas o velocidades del viento, para mejorar la predicción meteorológica.

5. **Investigación de Operaciones:** Utilizar distribuciones ajustadas en modelos de simulación para analizar y optimizar sistemas complejos.

6. **Estudios de Mercado:** Comprender la distribución de variables relevantes en estudios de mercado para la toma de decisiones empresariales.

7. **Fiabilidad y Mantenimiento:** Modelar la vida útil de equipos y sistemas para planificar estrategias de mantenimiento y garantizar la confiabilidad.

En resumen, el ajuste de distribuciones es una herramienta versátil con aplicaciones en diversas disciplinas, proporcionando una manera efectiva de entender, modelar y utilizar datos en la toma de decisiones y análisis de diferentes fenómenos.

````
# Importar librerías
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# También puede leer el archivo desde el repositorio de GitHub.
data = pd.read_csv("https://raw.githubusercontent.com/marsgr6/EN-online/8a1cee296279c274d8f05f3905fce97c1f7636a2/data/DL_data.csv")

data.head()
````

````
# Extraer solo la columna (variable) a modelar
enlaces = data['Enlaces']

# Valores únicos observados para la variable y sus frecuencias
x_obs, f_obs = np.unique(enlaces, return_counts=True)
# Agregamos el 10 que no aparece en los datos con frecuencia 0
x_obs = np.insert(x_obs, -1, 10)  # número de conexiones (grado del nodo)
# insertamos 0 en f_obs
f_obs = np.insert(f_obs, -1, 0)

# Calculamos las frecuencias relativas ("probabilidad" observada)
px_obs = f_obs/f_obs.sum()
# Calculamos las frecuencias acumuladas
Fx_obs = px_obs.cumsum()

o_data = pd.DataFrame({'x_obs': x_obs, 'f_obs': f_obs, 'prob_f_rel': px_obs, 'Fx_obs': Fx_obs})
o_data
````

````
def plot_pmf(x, pmfx, ax=None):
    if ax is None: fig, ax = plt.subplots(1, 1)
    # https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.vlines.html
    ax.vlines(x, 0, pmfx, colors='r', linestyles='-', lw=1)  # líneas verticales
    ax.plot(x, pmfx, 'ro', label='Pmf')  # puntos sobre las lineas verticales
    for xi, fxi in zip(x, pmfx):
        ax.text(xi, fxi, str(fxi))  # Imprimimos el valor de fx (PMF)


def plot_cdf(rvs_range, cdf_prob, ax=None):
    yn = np.insert(cdf_prob, 0, 0)
    
    if ax is None:
        fig, ax = plt.subplots()
        ax.set_facecolor('white')

    # https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.hlines.html
    ax.hlines(y=yn, xmin=rvs_range[:-1], xmax=rvs_range[1:],
              color='red', zorder=1)

    # https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.vlines.html
    ax.vlines(x=rvs_range[1:-1], ymin=yn[:-1], ymax=yn[1:], color='red',
              linestyle='dotted', zorder=1)

    ax.scatter(rvs_range[1:-1], cdf_prob, color='red', s=18, zorder=2)
    ax.scatter(rvs_range[1:-1], yn[:-1], color='white', s=18, zorder=2,
               edgecolor='red')


fig, ax = plt.subplots(1, 2, figsize=(8,4))

plot_pmf(o_data.x_obs, o_data.f_obs, ax[0])
ax[0].set_title("PMF observada")
x = np.arange(x_obs.min()-1, x_obs.max()+2)  # Soporte, valores posibles de x +-1 para las colas de la distribución
plot_cdf(x, list(o_data.Fx_obs), ax=ax[1])

ax[1].set_title("CDF observada")
plt.show()
````

### Ejemplo

**Asumiremos una distribución de Poisson de acuerdo al enunciado**
- El parámetro $\lambda=5$, también nos viene dado por el enunciado.
  - En realidad desconocemos $\lambda$ y queremos calcularlo a partir de la muestra. 
  - Necesitamos calcular por tanto, el estimador de máxima verosimilitud (MLE) para $\lambda$.

En el caso de la distribución de Poisson, el estimador de máxima verosimilitud (MLE) para el parámetro $ \lambda $ (la tasa de eventos en un intervalo fijo de tiempo o espacio) se obtiene maximizando la función de verosimilitud.

La función de verosimilitud para una muestra $ x_1, x_2, \ldots, x_n $ de variables aleatorias Poisson con parámetro $ \lambda $ es dada por:

$$ L(\lambda) = \prod_{i=1}^{n} \frac{e^{-\lambda} \lambda^{x_i}}{x_i!} $$

Donde $ x_i $ es la observación $ i $-ésima en la muestra.

Para simplificar el cálculo, se suele trabajar con el logaritmo de la función de verosimilitud, conocido como el logaritmo de la verosimilitud ($ \log L(\lambda) $):

$$ \log L(\lambda) = \sum_{i=1}^{n} \left(-\lambda + x_i \log(\lambda) - \log(x_i!)\right) $$

Luego, para encontrar el estimador de máxima verosimilitud $ \hat{\lambda} $, derivamos $ \log L(\lambda) $ con respecto a $ \lambda $, igualamos a cero y resolvemos para $ \lambda $:

$$ \frac{d(\log L(\lambda))}{d\lambda} = -n + \frac{\sum_{i=1}^{n} x_i}{\lambda} = 0 $$

Resolviendo para $ \lambda $, obtenemos:

$$ \hat{\lambda} = \frac{\sum_{i=1}^{n} x_i}{n} $$

Por lo tanto, el estimador de máxima verosimilitud para $ \lambda $ en la distribución de Poisson es simplemente el promedio de las observaciones en la muestra.

````
# Importar distribución teórica
from scipy.stats import poisson

# lambda=5 de acuerdo al enunciado
l = 5  # lambda (parámetro de la dist. Poisson)

# Modelo: PMF
# dist.pmf(x, parametros)
px_model_l5 = poisson.pmf(x_obs, l)  # lambda=5

# Función de distribución acumulada CDF
# dist.cdf(x, parametros)
Fx_model_l5 = poisson.cdf(x_obs, l)  # el valor final loc=1, desplaza la distribución

# Modelo: PMF
# dist.pmf(x, parametros)
px_model_l3 = poisson.pmf(x_obs, 3)  # lambda=3
px_model_l4 = poisson.pmf(x_obs, 4)  # lambda=4
# Estimador de máxima verorsimilitud (MLE en inglés)
l_MLV = x_obs.mean()
px_model_l_MLV = poisson.pmf(x_obs, l_MLV)  # lambda=MLE

# Función de distribución acumulada CDF
# dist.cdf(x, parametros); el valor final loc=1, desplaza la distribución
Fx_model_l3 = poisson.cdf(x_obs, 3)  # lambda=3 
Fx_model_l4 = poisson.cdf(x_obs, 4)  # lambda=4
Fx_model_l_MLV = poisson.cdf(x_obs, l_MLV)  # lambda=MLE

# Graficamos
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)  # panel izquierdo
plt.bar(x_obs, px_obs, label='Prob obs')
plt.plot(x_obs, px_model_l3, ':og', label=r'PMF modelo $\lambda=3$')  # lambda=3
plt.plot(x_obs, px_model_l4, ':ok', label=r'PMF modelo $\lambda=4$')  # lambda=4
plt.plot(x_obs, px_model_l5, ':or', label=r'PMF modelo $\lambda=5$')  # lambda=5
plt.plot(x_obs, px_model_l_MLV, ':ob', label=r'PMF modelo MLE $\lambda=$'+str(l_MLV))  # lambda=MLE
plt.legend()

plt.subplot(1,2,2)  # panel derecho
plt.bar(x_obs, Fx_obs, label='Prob acum obs')
plt.plot(x_obs, Fx_model_l3, ':og', label=r'CDF modelo $\lambda=3$')  # lambda=3
plt.plot(x_obs, Fx_model_l4, ':ok', label=r'CDF modelo $\lambda=4$')  # lambda=4
plt.plot(x_obs, Fx_model_l5, ':or', label=r'CDF modelo $\lambda=5$')  # lambda=5
plt.plot(x_obs, Fx_model_l_MLV, ':ob', label=r'PMF modelo MLE $\lambda=$'+str(l_MLV))  # lambda=MLE
plt.legend()
````

### Poisson vs binomial

````
from scipy.stats import binom
n, p = 50, 0.1  # Binomial(n=50, p=0.1)
l = n*p  # Poisson(lambda=n*p=5)

x = x_obs
p_binom = binom.pmf(x, n, p)  # Binomial model
p_poisson = poisson.pmf(x, l)  # Poisson model

# Graficamos
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)  # panel izquierdo
plt.bar(o_data.x_obs, o_data.prob_f_rel, label='Prob obs')
plt.plot(x, p_binom, ':or', label='Binomial')
plt.plot(x, p_poisson, ':Pg', label='Poisson')
plt.legend()

plt.subplot(1,2,2)  # panel derecho
plt.bar(o_data.x_obs, o_data.Fx_obs, label='Prob acum obs')
plt.plot(x, p_binom.cumsum(), ':or', label='Binomial CDF')
plt.plot(x, p_poisson.cumsum(), ':Pg', label='Poisson CDF')
plt.legend()
````

### Prueba de bondad de ajuste

- Revisar: https://nbviewer.org/github/marsgr6/analitica-online/blob/main/statistical_tests.ipynb#chi2ba

````
from scipy.stats import chi2
l = 5
x_obs = np.concatenate(([0], x_obs))
observed, expected = np.concatenate(([0], f_obs)), poisson.pmf(x_obs, l)*len(data.Enlaces)

alpha = 0.05
k = len(observed)
df = k - 1

Q_p = ((observed - expected)**2 / expected).sum()
chi2_c = chi2.ppf(1-alpha, df)
decision = {True: "Reject H0", False: "Do not reject H0"}
pvalue = 1-chi2.cdf(Q_p, df)
print({"Q_p": Q_p, "chi2_critico": chi2_c, "p-value": pvalue,
       "decision": decision[Q_p > chi2_c]})
summary = pd.DataFrame({'enlaces': x_obs, 
                        'poisson_prob': poisson.pmf(x_obs, l), 
                        'observed': observed, 
                        'expected':expected})

plt.bar(summary.enlaces, summary.observed/summary.observed.sum(), label='Prob obs')
plt.plot(summary.enlaces, summary.poisson_prob, ':or', label='PMF modelo')

summary
````
