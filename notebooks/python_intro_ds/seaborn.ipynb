```python
# Por qu√© Seaborn?
# Ver: https://dsaber.com/2016/10/02/a-dramatic-tour-through-pythons-data-visualization-landscape-including-ggplot-and-altair/
# Fuentes y recursos
# https://elitedatascience.com/python-seaborn-tutorial
# https://www.kaggle.com/aschakra/iris-data-visualization-using-python
# https://seaborn.pydata.org/tutorial/distributions.html
# See categorical data analysis: https://seaborn.pydata.org/tutorial/categorical.html

import pandas as pd
import numpy as np

iris = pd.read_csv("https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/datasets/iris.csv")
iris.head(8)  # Primeros 8 registros
```

```python
del iris['Unnamed: 0']
iris.head()
```

```python
iris["Species"].value_counts()
```

```python
# Visualizar distribuciones:
# https://seaborn.pydata.org/tutorial/distributions.html
# Podemos cargar iris de sns
# iris = sns.load_dataset("iris")
# Tenga cuidado que los nombres de las variables cambian

import matplotlib.pyplot as plt
import seaborn as sns  # importamos modulo seaborn como sns
sns.set()  # Establece par√°metros del gr√°fico (por defecto)

# Histogramas 
b = np.linspace(0, 2.7, 20)  # Rango para los bines del histograma
g = sns.FacetGrid(iris, hue="Species")
g = g.map(sns.distplot, "Petal.Width", bins=b)
plt.legend()
plt.show()
```

```python
# Diagramas de densidad

b = np.linspace(0, 2.7, 20)  # Rango para los bines del histograma
g = sns.FacetGrid(iris, hue="Species")
g = g.map(sns.kdeplot, "Petal.Width")
plt.legend()
plt.show()
```

```python
# Boxplots
sns.set()
sns.boxplot(x="Species", y="Sepal.Length", data=iris)
plt.show()
```

```python
# Intercambiamos y por x
sns.boxplot(y="Species", x="Petal.Length", data=iris)
plt.show()
```

```python
# Diagramas de viol√≠n
sns.violinplot(x="Species", y="Petal.Length", data=iris)
plt.show()
```

```python
# Diagrama de dispersi√≥n
sns.lmplot('Petal.Width', 'Sepal.Length', data=iris, hue='Species', fit_reg=True)
plt.show()
```

```python
# Scatter plots para Species
# hue: Variable in data to map plot aspects to different colors
sns.pairplot(iris, hue="Species")
plt.show()
```

```python
sns.pairplot(iris, hue="Species", diag_kind="kde")
plt.show()
```

```python
# Ver: http://resources.rhoworld.com/blog/resourcebrcenter/rhos-blog/i-swarm-you-swarm-we-all-swarm-for-beeswarm-plots-0

sns.set(style="whitegrid", palette="muted")  # Establece par√°metros del gr√°fico

sns.swarmplot(x="Attribute", y="value", hue="Species", data=iris2)
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")

sns.set()

# Paired Density and Scatterplot Matrix
# Ver: http://seaborn.pydata.org/generated/seaborn.PairGrid.html
g = sns.PairGrid(iris, diag_sharey=False, hue="Species")  # guardamos gr√°fico en g
g.map_lower(sns.kdeplot)  # agregamos la parte inferior (density plot 2D)
g.map_upper(plt.scatter)  # agregamos la parte inferior (scatter plot)
g.map_diag(sns.kdeplot, lw=3)  # agregamos la diagonal (density plot)
plt.show()
```

```python
# Algo m√°s complejo
# https://stackoverflow.com/questions/32889590/seaborn-pairgrid-using-kdeplot-with-2-hues

def infer_cmap(color):  
    if color == (0., 0., 1.):
        return 'Blues'
    elif color == (0., 0.5, 0.):
        return 'Greens'
    elif color == (1., 0., 0.):
        return 'Reds'


def kde_hue(x, y, **kws):
    # ax = plt.gca()
    cmap = infer_cmap(kws['color'])
    return sns.kdeplot(data=x, data2=y, shade=True, shade_lowest=False, cmap=cmap, **kws)
    #return ax


colores = ['b', 'g', 'r']
var = 'Species'

color_dict = {}
for idx, v in enumerate(np.unique(iris[var])):
    color_dict[v] = colores[idx]
    
g = sns.PairGrid(iris, hue=var, palette=color_dict)
g = g.map_lower(kde_hue)
g = g.map_upper(plt.scatter, s=20)
g = g.map_diag(sns.kdeplot)
g = g.add_legend()
plt.show()
```
