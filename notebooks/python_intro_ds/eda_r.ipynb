```r
# Podemos acceder a los atributos del dataset usando el operador $
iris$Petal.Length
attach(iris) # Agrega el conjunto de datos a la ruta de búsqueda (search path) de R
Petal.Length
```

```r
# Histogramas en R
# Intente los siguientes comandos para entender la construcción de los histogramas
# iris$Sepal.Length>6
# iris$Species=="setosa"
# Las 2 líneas anteriores devuelven un valor lógico TRUE para las filas
# que cumplen con la condición, podemos extraer por lo tanto aquellas
# filas que tienen condición verdadera (TRUE)
p1 <- hist(iris$Petal.Width[unclass(iris$Species=="setosa")], breaks=seq(0,2.75,by=0.1), plot=F)
p2 <- hist(iris$Petal.Width[unclass(iris$Species=="versicolor")], breaks=seq(0,2.75,by=0.1), plot=F)
p3 <- hist(iris$Petal.Width[unclass(iris$Species=="virginica")], breaks=seq(0,2.75,by=0.1), plot=F)
# El comando breaks indica el número de bines que será igual para todos los casos (Species)
# Note el uso del comando seq

# Graficamos los histogramas construidos anteriormente
# Indicamos las etiquetas de los ejes y del gráfico
# así como el color usado
# Definimos ejes iguales para todos los histogramas
# add=T, permite agregar el gráfico a la misma ventana
plot(p1, col="red",xlim=c(0,3), main='Iris Data', xlab="Petal Width", ylab='Frecuencia')
plot(p2, col="green",xlim=c(0,3), main='Iris Data', xlab="Petal Width", ylab='Frecuencia', add=T)
plot(p3, col="blue",xlim=c(0,3), main='Iris Data', xlab="Petal Width", ylab='Frecuencia', add=T)

# Legenda para indicar los colores
# legend("topright", c("Setosa","Virginica","Versicolor"), col=c("red", "green", "blue"), pch=15, cex=1.5)
legend("topright", c(levels(iris$Species)), col=c("red", "green", "blue"), pch=15, cex=1.5)
```

```r
# Diagramas de densidad (density plots)
# par() Establece los parámetros graicos, por ejemplo
# permite combinar múltiples gráficos en R
# Ver: https://stat.ethz.ch/R-manual/R-devel/library/graphics/html/par.html
d1 <- density(iris$Petal.Width[unclass(iris$Species=="setosa")])
d2 <- density(iris$Petal.Width[unclass(iris$Species=="versicolor")])
d3 <- density(iris$Petal.Width[unclass(iris$Species=="virginica")])
# Tenga cuidad con las escalas (xlim, ylim), deben ser las mismas para los 3 gráficos
plot(d1, main="Iris Data", xlab="Petal Width", col="red", xlim=c(0,3), ylim=c(0,8), lwd=2)
par(new=TRUE)
plot(d2, main="Iris Data", xlab="Petal Width", col="green", xlim=c(0,3), ylim=c(0,8), lwd=2)
par(new=TRUE)
plot(d3, main="Iris Data", xlab="Petal Width", col="blue", xlim=c(0,3), ylim=c(0,8), lwd=2)
legend("topright", c(levels(iris$Species)), col=c("red", "green", "blue"), lty=1, lw=2, cex=1.5)
```

```r
# Boxplots son simples de construir
# ~ es el operador formula en R
# Aquí indica que Petal.Width depende de iris
boxplot(Petal.Width~Species,data = iris,col=c("red", "green","blue"), horizontal=T, main="Iris Data")
```

```r
# Scatter plots
# attach(iris) # Debe hacer un attach de iris, si no lo ha hecho antes
# plot(Petal.Length, Petal.Width, main="Iris Data")
# plot(Petal.Length, Petal.Width, pch=c(23,24,25)[unclass(Species)], main="Iris Data", xlab="Petal Length", ylab="Petal Width")
# plot(Petal.Length, Petal.Width, pch=21, bg=c("red","green3","blue")[unclass(Species)], main="Iris Data", xlab="Petal Length", ylab="Petal Width")


## put histograms on the diagonal
panel.cor <- function(x, y, digits = 2, prefix = "", cex.cor, ...)
 {
     usr <- par("usr"); on.exit(par(usr))
     par(usr = c(0, 1, 0, 1))
     r <- abs(cor(x, y))
     txt <- format(c(r, 0.123456789), digits = digits)[1]
     txt <- paste0(prefix, txt)
     if(missing(cex.cor)) cex.cor <- 0.8/strwidth(txt)
     text(0.5, 0.5, txt, cex = cex.cor * r)
 }

## put (absolute) correlations on the upper panels,
## with size proportional to the correlations.
panel.hist <- function(x, ...)
 {
     usr <- par("usr"); on.exit(par(usr))
     par(usr = c(usr[1:2], 0, 1.5) )
     h <- hist(x, plot = FALSE)
     breaks <- h$breaks; nB <- length(breaks)
     y <- h$counts; y <- y/max(y)
     rect(breaks[-nB], 0, breaks[-1], y, col = "cyan", ...)
 }

pairs(iris[1:4], main = "Iris Data", pch = 21, 
      bg = c("red", "green3", "blue")[unclass(Species)])
    
pairs(iris[1:4], main = "Iris Data", pch = 21, 
      bg = c("red", "green3", "blue")[unclass(Species)], diag.panel=panel.hist, upper.panel=panel.cor)
```
