```python
# Gráficos con matplotlib

# Fuentes y Recursos
# https://matplotlib.org/users/screenshots.html
# http://www.scipy-lectures.org/intro/matplotlib/matplotlib.html

import numpy as np
import matplotlib.pyplot as plt

angulo = np.linspace(0, 2*np.pi, 100)  # 100 valores entre 0 y 2pi

seno = np.sin(angulo)  # evaluamos el seno
coseno = np.cos(angulo)  # evaluamos coseno

plt.plot(angulo, seno)  # graficamos 
plt.plot(angulo, coseno)

plt.legend((r"$\sin(\theta)$", r"$\cos(\theta)$"))  # leyendas
plt.title("Trignometric Functions")  # título
plt.xlabel(r"$\theta$ (radians)")  # etiqueta para eje x
plt.ylabel(r"$f(\theta)$")  # etiqueta para eje y

plt.show()  # mostrar gráfico
```

```python
# Subplots
# plt.subplot(nrows, ncols, plot_number) 

plt.subplot(2, 1, 1)  # 2 
plt.plot(angulo, seno, '--', color='b')
plt.title('Trignometric Functions')
plt.ylabel(r'$\sin(\theta$)')

plt.subplot(2, 1, 2)
plt.plot(angulo, coseno, ':', color='g')
plt.xlabel(r'$\theta$')
plt.ylabel(r'$\cos(\theta$) (radians)')

plt.show()
```

```python
fig, ax = plt.subplots(2, 1)  # subplots devuelve un objeto figura

ax[0].plot(angulo, seno, color='b', label="Seno")
ax[0].set_ylabel(r'$\sin(\theta$)')
ax[0].set_title("Trigonometric functions")

ax[1].plot(angulo, coseno, color='b', label="Coseno")
ax[1].set_ylabel(r'$\cos(\theta$)')
ax[1].set_xlabel(r'$\theta$ (radians))')
ax[1].grid(True, linestyle=':')
ax[1].set_xticks(np.arange(0, 7, 0.5))
ax[1].set_yticks(np.linspace(-1, 1, 5))
    
plt.show()
```

```python
# Subplots
# https://en.wikipedia.org/wiki/Damped_sine_wave

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('Oscillations')
plt.ylabel('Damped')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()

# Vea: https://matplotlib.org/api/lines_api.html
# para tipos de línea y marcadores
```

```python
# Diagrama de barras
means_men = (20, 35, 30, 35, 27)  # Frecuencias
bar_width = 0.35  # ancho de la barra
opacity = 0.4  # Opacidad de las barras

n_groups = 5

index = np.arange(n_groups)

plt.bar(index, means_men, bar_width,
        alpha=opacity,
        color='b',
        label='Men')


plt.title('Scores by group and gender')
plt.ylabel('Scores')
plt.xticks(index, ('A', 'B', 'C', 'D', 'E'))
plt.legend()
plt.show()
```

```python
"""
Bar chart demo with pairs of bars grouped for easy comparison.
"""
import numpy as np
import matplotlib.pyplot as plt


n_groups = 5

means_men = (20, 35, 30, 35, 27)
std_men = (2, 3, 4, 1, 2)

means_women = (25, 32, 34, 20, 25)
std_women = (3, 5, 2, 3, 3)

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.35

opacity = 0.4

# Barras de error, color 
error_config = {'ecolor': 'black', 'capsize': 3}

rects1 = plt.bar(index, means_men, bar_width,
                 alpha=opacity,
                 color='b',
                 yerr=std_men,
                 error_kw=error_config,
                 label='Men')

rects2 = plt.bar(index + bar_width, means_women, bar_width,
                 alpha=opacity,
                 color='r',
                 yerr=std_women,  # barras de error en y
                 error_kw=error_config,
                 label='Women')

plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(index + bar_width / 2, ('A', 'B', 'C', 'D', 'E'))
plt.legend()

# tight_layout automatically adjusts subplot params so that 
# the subplot(s) fits in to the figure area. 
plt.tight_layout()
plt.show()

# Ver: https://matplotlib.org/users/screenshots.html
```

```python
# Line plot with error bars

plt.errorbar(
    index,   # X
    means_men,  # Y
    yerr=std_men,  # Y-errors
    label="Men",
    fmt="rs--",  # format line like for plot()
    capsize = 2,
    linewidth=2  # width of plot line
    )
plt.legend() #Show legend
plt.show()
```

```python
# Scatter plots

n = 1024  # número de puntos a generar
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)
T = np.arctan2(Y, X) # https://en.wikipedia.org/wiki/Atan2

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.scatter(X, Y, s=15, c=T, alpha=.5)

plt.xlim(-3, 3)
# plt.xticks(())
plt.ylim(-3, 3)
# plt.yticks(())

plt.show()
```

```python
# Histograma 2D 
plt.hist2d(X, Y, bins=50)
cbar = plt.colorbar()
cbar.ax.set_ylabel('Counts')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd

# Descargar datos de:
# https://drive.google.com/file/d/1BEoom36awM318Aan2VoXTrHtsqU65Q_g/view?usp=sharing

pc_data = pd.read_csv("pca_data.csv")

pc_data.head()
```

```python
x = pc_data['Comp.1']
y = pc_data['Comp.2']
z = pc_data['Comp.3']
col = np.where(pc_data['cluster']==1, 'b', 'r')
mc = np.where(pc_data['cluster']==1, 'o', '^')

fig = plt.figure()
ax3D = fig.add_subplot(111, projection='3d')

for i in range(30):
    xs = x[i]
    ys = y[i]
    zs = z[i]
    ax3D.scatter(xs, ys, zs, s=90, c=col[i], marker=mc[i])


ax3D.tick_params(axis='both', which='major', labelsize=15)
ax3D.xaxis.label.set_size(20)
ax3D.yaxis.label.set_size(20)
ax3D.zaxis.label.set_size(20)

ax3D.set_xlim([-2.5, 2.5])
ax3D.set_ylim([-2.5, 2.5])
ax3D.set_zlim([-2.5, 2.5])

ax3D.set_xlabel('Component 1')
ax3D.set_ylabel('Component 2')
ax3D.set_zlabel('Component 3')

plt.show()
```
