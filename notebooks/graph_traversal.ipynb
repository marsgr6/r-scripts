### Graphs 

Graphs are mathematical structures that consist of a set of vertices (also known as nodes) and a set of edges (also known as connections) that connect pairs of vertices. Graphs are widely used to represent relationships between objects or entities in various domains, such as computer networks, social networks, transportation networks, and more.

Graph optimization algorithms are algorithms that aim to solve optimization problems on graphs. These algorithms leverage the structure of the graph to find the best solution or optimize a specific objective function. Some common graph optimization algorithms include:

1. Shortest Path Algorithms: These algorithms find the shortest path between two vertices in a graph. Examples include Dijkstra's algorithm and Bellman-Ford algorithm.

2. Minimum Spanning Tree Algorithms: These algorithms find a tree that spans all the vertices of a graph with the minimum total edge weight. Examples include Kruskal's algorithm and Prim's algorithm.

3. Maximum Flow Algorithms: These algorithms find the maximum flow that can be sent through a network of interconnected vertices. Examples include Ford-Fulkerson algorithm and Edmonds-Karp algorithm.

4. Traveling Salesman Problem Algorithms: These algorithms aim to find the shortest possible route that visits a set of given vertices exactly once and returns to the starting vertex. Examples include the Held-Karp algorithm and the Lin-Kernighan algorithm.

5. Graph Coloring Algorithms: These algorithms aim to assign colors to the vertices of a graph such that no two adjacent vertices have the same color. Examples include the Greedy algorithm and the Backtracking algorithm.

6. Clustering Algorithms: These algorithms aim to partition the vertices of a graph into clusters or communities based on their connectivity. Examples include the Girvan-Newman algorithm and the Louvain algorithm.

These are just a few examples of graph optimization algorithms. There are many more algorithms that address different optimization problems on graphs, each with its own characteristics and applications.

### Graph search and traversal

Blind search, also known as uninformed search, is a class of graph traversal algorithms that explore a graph without any knowledge about the goal or the structure of the graph. These algorithms are called "blind" because they do not have any information to guide their search other than the connectivity of the graph.

Graph traversal algorithms, on the other hand, are techniques used to visit all the nodes or vertices of a graph. These algorithms are used to explore or search a graph in a systematic manner. They are commonly used in various applications such as pathfinding, network analysis, and optimization problems.

There are several types of graph traversal algorithms, including depth-first search (DFS) and breadth-first search (BFS). Let's take a closer look at these two algorithms:

1. Depth-First Search (DFS): DFS starts at a given node and explores as far as possible along each branch before backtracking. It uses a stack data structure to keep track of the nodes to visit. DFS is often implemented recursively, where each recursive call explores a neighboring node until there are no more unvisited nodes. DFS is useful for exploring deep into a graph and can be used to find paths or cycles.

2. Breadth-First Search (BFS): BFS explores all the neighboring nodes of a given node before moving on to the next level of nodes. It uses a queue data structure to keep track of the nodes to visit. BFS visits nodes in a level-by-level manner, starting from the initial node. BFS is useful for finding the shortest path between two nodes or for exploring a graph in a systematic manner.

Here's an example to illustrate the use of graph traversal algorithms:

Suppose we have a graph representing a city's road network, where each node represents an intersection and each edge represents a road connecting two intersections. We want to find the shortest path from one intersection to another.

We can use a graph traversal algorithm like BFS to explore the road network starting from the initial intersection. BFS will visit neighboring intersections first, then move on to the next level of intersections until it reaches the destination intersection. By keeping track of the visited intersections and the path taken, we can find the shortest path between the two intersections.

In summary, blind search refers to graph traversal algorithms that explore a graph without any prior knowledge. Graph traversal algorithms like DFS and BFS are used to systematically explore a graph and can be applied to various problems such as pathfinding and network analysis.

### Trees



Trees and graphs are both data structures used to represent relationships between objects. However, there are some key differences between them.

A tree is a type of graph that is characterized by its hierarchical structure. It consists of nodes connected by edges, where each node (except the root node) has exactly one parent node and zero or more child nodes. The root node is the topmost node in the tree, and it has no parent. The leaf nodes are the nodes at the bottom of the tree that have no children. Trees are often used to represent hierarchical relationships, such as file systems, organization charts, or family trees.

Some important properties of trees include:

1. Uniqueness of Paths: There is a unique path between any two nodes in a tree.
2. Acyclic Structure: A tree does not contain any cycles or loops.
3. Connectedness: All nodes in a tree are connected.

On the other hand, a graph is a more general data structure that consists of a set of vertices (nodes) connected by edges. Unlike trees, graphs can have cycles and multiple edges between nodes. Graphs are used to represent various types of relationships, such as social networks, transportation networks, or computer networks.

Some important properties of graphs include:

1. Cyclic Structure: Graphs can have cycles or loops.
2. Multiple Edges: Graphs can have multiple edges between the same pair of nodes.
3. Disconnectedness: Graphs can have disconnected components, where some nodes are not reachable from others.

In summary, trees are a specific type of graph that have a hierarchical structure, while graphs are more general and can have cycles and multiple edges. Trees are often used to represent hierarchical relationships, while graphs are used to represent various types of relationships and connections.

The Breadth-First Search (BFS) algorithm is used to traverse or search a graph or a tree. While the basic principles of BFS remain the same for both trees and graphs, there are some differences in how the algorithm is applied to each.

1. Data Structure: In BFS for a tree, a queue is typically used to store the nodes to be visited. The algorithm starts at the root node and explores its children in a level-by-level manner. On the other hand, BFS for a graph can also use a queue, but it requires additional data structures to keep track of visited nodes and to handle cycles in the graph.

2. Connectivity: Trees are inherently connected structures, meaning that there is only one path between any two nodes. In BFS for a tree, there is no need to keep track of visited nodes because there are no cycles or multiple paths to consider. In contrast, graphs can have cycles and multiple paths between nodes. Therefore, BFS for a graph requires keeping track of visited nodes to avoid revisiting them and getting stuck in an infinite loop.

3. Terminating Condition: In BFS for a tree, the algorithm terminates when all nodes have been visited, i.e., the entire tree has been traversed. In BFS for a graph, the termination condition depends on the specific problem being solved. For example, the algorithm may terminate when a specific node or condition is reached, or when all reachable nodes have been visited.

4. Time Complexity: The time complexity of BFS for both trees and graphs is generally O(V + E), where V is the number of vertices (nodes) and E is the number of edges. However, in the case of trees, the number of edges is typically V-1, resulting in a slightly lower time complexity of O(V).

Overall, while the basic concept of BFS remains the same for trees and graphs, the implementation details and considerations differ due to the presence of cycles and multiple paths in graphs.

### BFS (Breadth-First Search)

BFS (Breadth-First Search) is a graph traversal algorithm that explores all the vertices of a graph in breadth-first order, i.e., it visits all the vertices at the same level before moving to the next level. It starts at a given source vertex and explores all its neighbors before moving to the next level of vertices.

BFS Pseudocode:
```
BFS(graph, start_vertex):
    create an empty queue
    enqueue start_vertex into the queue
    mark start_vertex as visited
    
    while the queue is not empty:
        dequeue a vertex from the queue
        process the dequeued vertex
        
        for each neighbor of the dequeued vertex:
            if the neighbor is not visited:
                enqueue the neighbor into the queue
                mark the neighbor as visited
```

Characteristics of BFS:
- BFS explores all vertices at the same level before moving to the next level.
- It uses a queue data structure to keep track of the vertices to be visited.
- BFS guarantees that it visits all vertices reachable from the source vertex.
- It finds the shortest path between the source vertex and any other vertex in an unweighted graph.
- BFS can be used to detect cycles in a graph.
- The time complexity of BFS is O(V + E), where V is the number of vertices and E is the number of edges in the graph.

Cost of BFS:
- The space complexity of BFS is O(V), where V is the number of vertices in the graph. This is because it needs to store the visited vertices and the vertices in the queue.
- The time complexity of BFS is O(V + E), where V is the number of vertices and E is the number of edges in the graph. This is because it needs to visit each vertex and each edge once.

In summary, BFS is an efficient algorithm for traversing or searching in a graph, especially when the graph is unweighted or the shortest path is required. It explores the graph in a breadth-first manner, visiting all vertices at the same level before moving to the next level.

### Depth-First Search

DFS stands for Depth-First Search, which is a graph traversal algorithm used to explore or search through a graph or tree data structure. It starts at a given node and explores as far as possible along each branch before backtracking.

Here is the pseudocode for the DFS algorithm:

```
DFS(node):
    if node is not visited:
        mark node as visited
        for each neighbor of node:
            if neighbor is not visited:
                DFS(neighbor)
```

Characteristics of DFS:

1. Depth-First Search explores a graph or tree in a depthward motion before visiting its siblings or neighbors.
2. It uses a stack data structure to keep track of the nodes to visit.
3. DFS is often implemented using recursion, where each recursive call explores a new node.
4. It is used to solve problems like finding connected components, detecting cycles, and solving mazes.
5. DFS can be used for both directed and undirected graphs.

### Dijkstra's algorithm 

Dijkstra's algorithm is a popular algorithm used to find the shortest path between two nodes in a graph. It works on both weighted and unweighted graphs, but it is most commonly used for weighted graphs.

The algorithm starts at a given source node and explores the neighboring nodes, calculating the shortest distance from the source to each node. It maintains a priority queue of nodes, prioritized by their current shortest distance from the source. The algorithm continues until all nodes have been visited or until the destination node is reached.

Here is a pseudocode representation of Dijkstra's algorithm:

```
function Dijkstra(Graph, source):
    create empty set S
    create a set Q of all nodes in Graph
    initialize distances to all nodes as infinity except for the source node, which is 0
    while Q is not empty:
        current = node in Q with the smallest distance
        remove current from Q
        add current to S
        for each neighbor of current:
            if neighbor is not in S:
                calculate new distance from source to neighbor
                if new distance is less than current distance:
                    update neighbor's distance
    return distances
```



Characteristics of Dijkstra's algorithm:
1. It guarantees finding the shortest path in a graph with non-negative edge weights.
2. It can handle both directed and undirected graphs.
3. It uses a greedy approach by selecting the node with the smallest distance at each step.
4. It maintains a priority queue to efficiently select the next node to visit.
5. It has a time complexity of O((V + E) log V), where V is the number of vertices and E is the number of edges in the graph.

### Minimum spanning tree

A minimum spanning tree (MST) is a subset of the edges of a connected, weighted graph that connects all the vertices together without any cycles and has the minimum possible total edge weight. It is often used to find the most efficient way to connect a set of nodes or locations.

Here is a pseudo code for finding a minimum spanning tree using Kruskal's algorithm:

1. Sort all the edges of the graph in non-decreasing order of their weights.
2. Create an empty set to store the minimum spanning tree.
3. Iterate through the sorted edges:
     - If adding the current edge to the minimum spanning tree does not create a cycle, add it to the tree.
     - Otherwise, discard the edge.
4. Return the minimum spanning tree.

Characteristics of a minimum spanning tree include:

1. It is a tree: A minimum spanning tree is a connected acyclic graph that spans all the vertices of the original graph.

2. It has the minimum total weight: The sum of the weights of the edges in a minimum spanning tree is minimized. This ensures that the tree connects all the vertices while minimizing the overall cost.

3. It is unique (in some cases): In some cases, there may be multiple minimum spanning trees with the same total weight. However, in many cases, the minimum spanning tree is unique.

### Network definition 

A graph or network can be defined mathematically as a set of nodes or vertices connected by edges. Mathematically, a graph can be represented as G = (V, E), where V represents the set of nodes or vertices, and E represents the set of edges connecting the nodes.

The set of nodes, V, can be finite or infinite, and each node can be labeled or identified uniquely. The set of edges, E, represents the connections or relationships between the nodes. Each edge can be represented as a pair of nodes (u, v), where u and v are nodes in V.

In addition to the basic mathematical representation, a graph can also have additional attributes or properties associated with the nodes or edges. For example, nodes can have labels, weights, or other attributes, and edges can have directions, weights, or other attributes.

Graphs can be classified into different types based on their properties and characteristics. Some common types of graphs include:

1. Undirected Graph: In an undirected graph, the edges do not have a specific direction. The connection between nodes is bidirectional.

2. Directed Graph: In a directed graph, also known as a digraph, the edges have a specific direction. The connection between nodes is unidirectional.

3. Weighted Graph: In a weighted graph, each edge is assigned a weight or value. The weights can represent distances, costs, or any other relevant measure.

4. Bipartite Graph: In a bipartite graph, the nodes can be divided into two disjoint sets, and edges only connect nodes from different sets.

5. Complete Graph: In a complete graph, every pair of distinct nodes is connected by an edge. It is a fully connected graph.

These are just a few examples of the types of graphs that can be defined mathematically. Graph theory provides a rich set of mathematical tools and concepts to analyze and study the properties and behavior of graphs in various domains.

### Network Analysis and Statistics

Now, let's discuss the statistics that can be calculated at different levels of the network:

1. Microscopic Level:
    - Degree Distribution: The distribution of node degrees in the network, which represents the number of connections each node has.
    - Clustering Coefficient: Measures the tendency of nodes to form clusters or groups.
    - Node Centrality Measures: Measures the importance or influence of individual nodes, such as degree centrality, betweenness centrality, and closeness centrality.

2. Mesoscopic Level:
    - Community Detection: Identifies groups or communities of nodes that are densely connected within themselves but sparsely connected with other communities.
    - Modularity: Measures the strength of the division of a network into communities.
    - Assortativity: Measures the tendency of nodes to connect with nodes of similar degree.

3. Macroscopic Level:
    - Network Diameter: The maximum shortest path length between any two nodes in the network, which represents the overall size of the network.
    - Average Path Length: The average shortest path length between all pairs of nodes in the network, which represents the average distance between nodes.
    - Network Density: The ratio of the number of edges to the maximum possible number of edges, which represents the connectedness of the network.

### Connectivity models

Connectivity models in graph theory refer to mathematical models that describe the connections or relationships between nodes in a graph. These models help analyze and understand the structure and properties of graphs.

### Complete graphs

A complete graph is a specific type of connectivity model where every pair of distinct nodes in the graph is connected by a unique edge. In other words, in a complete graph, there is an edge between every pair of nodes.

Properties of a complete graph:
1. Number of edges: In a complete graph with n nodes, the total number of edges is given by n(n-1)/2. This is because each node is connected to every other node except itself.
2. Degree of nodes: In a complete graph, each node has a degree of n-1, where n is the total number of nodes. This means that each node is connected to every other node in the graph.
3. Symmetry: The edges in a complete graph are symmetric, meaning that if there is an edge between node A and node B, there is also an edge between node B and node A.
4. Dense connectivity: A complete graph is considered to be maximally connected since every pair of nodes is directly connected by an edge. This makes it a useful model for studying connectivity and communication in networks.

Complete graphs are often represented by the symbol Kn, where n represents the number of nodes in the graph. For example, K3 represents a complete graph with 3 nodes, and K5 represents a complete graph with 5 nodes.

Complete graphs have various applications in graph theory, network analysis, and optimization problems. They serve as a benchmark for comparing the connectivity and efficiency of other graphs and algorithms.


### Erdos-Renyi network

An Erdos-Renyi network, also known as a random graph, is a mathematical model for generating random networks. It was introduced by Paul Erdos and Alfred Renyi in 1959. 

In an Erdos-Renyi network, the construction process involves randomly connecting nodes with edges based on a set of parameters. The two main parameters of an Erdos-Renyi network are:

1. Number of Nodes (n): It represents the total number of nodes or vertices in the network. Each node is typically represented by a unique identifier.

2. Edge Probability (p): It represents the probability of an edge existing between any pair of nodes. Each pair of nodes is considered independently, and an edge is created between them with a probability of p.

The construction process of an Erdos-Renyi network can be summarized as follows:

1. Start with an empty graph with n nodes.

2. For each pair of nodes (i, j), where i < j:
    - Generate a random number between 0 and 1.
    - If the generated number is less than or equal to p, create an edge between nodes i and j.

3. Repeat step 2 for all possible pairs of nodes.

The resulting network is a random graph where the presence of each edge is determined independently based on the edge probability p. The resulting network can have a varying number of edges, depending on the value of p.

Erdos-Renyi networks are useful for studying the properties of random networks and serve as a baseline for comparison with other types of networks. They have been extensively studied in the field of network science and have provided insights into various phenomena, such as phase transitions, connectivity, and percolation.

![](https://upload.wikimedia.org/wikipedia/commons/thumb/1/13/Erdos_generated_network-p0.01.jpg/1920px-Erdos_generated_network-p0.01.jpg)

The probability of having a connected network in an Erdos-Renyi model depends on the number of nodes ($n$) and the edge probability ($p$). In general, as the number of nodes increases, the probability of having a connected network decreases.

The critical value for the edge probability ($p_c$) in an Erdos-Renyi model is given by the equation:

$$p_c = ln(n) / n$$

If the edge probability (p) is greater than the critical value ($p_c$), the network is likely to be connected. Conversely, if the edge probability is less than the critical value, the network is likely to be disconnected.

It's important to note that this critical value is an approximation and may vary depending on the specific characteristics of the network. Additionally, other factors such as the average degree of the network can also influence the probability of connectivity.

To ensure a connected network in an Erdos-Renyi model, it is recommended to set the edge probability ($p$) above the critical value ($p_c$).

The degree distribution of an Erdos-Renyi network follows a Poisson distribution. 

In an Erdos-Renyi network, each pair of nodes has a fixed probability of being connected by an edge. This probability is denoted as $p$. The degree of a node in this network represents the number of edges connected to that node.

The Poisson distribution is a probability distribution that describes the number of events occurring in a fixed interval of time or space, given a known average rate of occurrence. It is often used to model random events that occur independently of each other.

In the context of an Erdos-Renyi network, the Poisson distribution describes the probability of a node having a certain degree. The average degree of the network is denoted as $k$, and it is equal to the product of the total number of nodes $N$ and the edge probability $p$ ($k = N * p$).

The probability mass function of the Poisson distribution is given by the formula:

$$P(k) = (e^{-k} * k^k) / k!$$

Where:
- $P(k)$ is the probability of having exactly $k$ edges connected to a node.
- $e$ is the base of the natural logarithm (approximately 2.71828).

The Poisson distribution is characterized by its mean, which is equal to the average degree $k$ in the case of an Erdos-Renyi network. The variance of the distribution is also equal to the mean.

It's important to note that the Poisson distribution is an approximation for large networks with a low edge probability $p$. In practice, the degree distribution of real-world networks may deviate from the Poisson distribution due to various factors such as network structure and node preferences.

### Small-world network 

![](https://media.springernature.com/full/springer-static/image/art%3A10.1038%2F30918/MediaObjects/41586_1998_Article_BF30918_Fig1_HTML.gif)

A small-world network is a type of network characterized by two main properties: high clustering coefficient and short average path length. It is built using the Watts-Strogatz model, which involves rewiring a regular lattice network.

The parameters of a small-world network are:

1. Number of Nodes (N): It represents the total number of nodes or vertices in the network. Each node is typically represented by a unique identifier.

2. Average Degree (K): It represents the average number of edges connected to each node in the network. The degree of a node represents the number of edges connected to that node.

3. Rewiring Probability (p): It represents the probability of rewiring an edge in the network. Each edge is considered independently, and with a probability of p, it is rewired to a randomly chosen node.

The construction process of a small-world network can be summarized as follows:

1. Start with a regular lattice network, where each node is connected to its K nearest neighbors. The network has a high clustering coefficient but a long average path length.

2. For each edge in the network, generate a random number between 0 and 1.

3. If the generated number is less than or equal to the rewiring probability p, rewire the edge to a randomly chosen node in the network. This step introduces randomness and creates shortcuts in the network.

4. Repeat steps 2 and 3 for all edges in the network.

The resulting network is a small-world network with a high clustering coefficient and a short average path length. The rewiring process introduces randomness, creating shortcuts that reduce the average path length while preserving the local clustering.

Mathematically, the average path length (L) of a small-world network can be calculated as:

$$L = \frac{1}{N(N-1)} \sum_{i \neq j} d(i,j)$$

where N is the number of nodes, and d(i,j) is the shortest path length between nodes i and j.

The clustering coefficient (C) of a small-world network can be calculated as:

$$C = \frac{3 \times \text{number of triangles}}{\text{number of connected triples}}$$

where a triangle is a set of three nodes that are all connected to each other, and a connected triple is a set of three nodes where at least two of the three edges are present.

Small-world networks exhibit the "small-world phenomenon," where the average path length is significantly shorter than that of a random network, while maintaining a high clustering coefficient similar to that of a regular lattice network.

### Scale-free networks

A Barabasi-Albert network is a type of scale-free network that is characterized by the preferential attachment mechanism. It was introduced by Albert-László Barabási and Réka Albert in 1999.

In a Barabasi-Albert network, the construction process involves the following parameters:

1. Number of Nodes (N): It represents the total number of nodes or vertices in the network. Each node is typically represented by a unique identifier.

2. Initial Nodes (m0): It represents the number of nodes in the initial network. The initial network is typically a complete graph or a small regular graph.

3. Attachment Parameter (m): It represents the number of edges that are added to the network for each new node. The attachment parameter determines the preferential attachment mechanism, where nodes with a higher degree are more likely to receive new edges.

The construction process of a Barabasi-Albert network can be summarized as follows:

1. Start with an initial network of m0 nodes. Each node is connected to every other node in the initial network.

2. Add new nodes to the network one by one, until the total number of nodes reaches N.

3. For each new node, connect it to m existing nodes in the network. The probability of connecting to a particular node is proportional to its degree.

4. Repeat step 3 for all new nodes.

The resulting network is a scale-free network, where the degree distribution follows a power-law distribution. This means that a few nodes have a significantly higher degree compared to the majority of nodes in the network.

Mathematically, the degree distribution of a Barabasi-Albert network can be represented as:

$$P(k) \propto k^{-\gamma}$$

where P(k) is the probability of a node having degree k, and γ is the power-law exponent. Typically, the value of γ falls between 2 and 3.

The preferential attachment mechanism in a Barabasi-Albert network leads to the "rich-get-richer" phenomenon, where nodes with a higher degree are more likely to receive new edges. This mechanism results in the formation of hubs, which are highly connected nodes in the network.

In summary, a Barabasi-Albert network is a scale-free network that is built using the preferential attachment mechanism. It exhibits a power-law degree distribution, with a few highly connected nodes (hubs) and many nodes with lower degrees.

### Community detection

Community detection in graph refers to the process of identifying groups or communities of nodes that are densely connected within themselves and sparsely connected with nodes outside the community. It is an important task in graph analysis and has various applications in different domains.

The importance of community detection in graph can be summarized as follows:

1. Understanding network structure: Community detection helps in understanding the underlying structure and organization of complex networks. By identifying communities, we can gain insights into the modular nature of networks and the relationships between nodes.

2. Identifying functional units: Communities often represent functional units or cohesive groups within a network. By detecting communities, we can identify groups of nodes that perform similar functions or have similar roles within the network.

3. Predictive modeling: Community detection can be used as a feature in predictive modeling tasks. Communities can provide additional information about the relationships between nodes, which can improve the accuracy of predictive models.

4. Network visualization: Communities can be visualized as distinct groups within a network, making it easier to interpret and analyze large-scale networks. Visualizing communities can reveal patterns, clusters, and relationships that may not be apparent in the raw network.

5. Social network analysis: Community detection is widely used in social network analysis to identify groups of individuals with similar interests, affiliations, or social connections. It helps in understanding social structures, information diffusion, and influence dynamics within social networks.

6. Recommendation systems: Community detection can be used in recommendation systems to identify communities of users with similar preferences or interests. By recommending items or content to users within the same community, personalized recommendations can be provided.

7. Biological networks: Community detection is applied in biological networks to identify functional modules or protein complexes. It helps in understanding the organization and interactions within biological systems, such as protein-protein interaction networks or gene regulatory networks.

8. Fraud detection: Community detection can be used in fraud detection to identify groups of entities or individuals involved in fraudulent activities. By detecting communities of suspicious behavior, fraud patterns can be identified and preventive measures can be taken.

Overall, community detection in graph plays a crucial role in understanding network structure, identifying functional units, improving predictive modeling, facilitating network visualization, analyzing social networks, enhancing recommendation systems, studying biological networks, and detecting fraudulent activities. It is a powerful tool for analyzing and interpreting complex networks in various domains.

````python
import networkx as nx

import matplotlib.pyplot as plt

# Create an empty graph
G = nx.Graph()

# Define the width and height of the grid
width = 20
height = 4

# Add nodes to the graph with positions as their names
for x in range(width + 1):
    for y in range(height + 1):
        G.add_node((x, y), pos=(x, y))

# Add edges to connect the nodes
for x in range(width + 1):
    for y in range(height + 1):
        # Connect with the top-right diagonal neighbor
        if x < width and y < height:
            G.add_edge((x, y), (x + 1, y + 1))
        # Connect with the top-left diagonal neighbor
        if x > 0 and y < height:
            G.add_edge((x, y), (x - 1, y + 1))

# Get the positions of the nodes
pos = nx.get_node_attributes(G, 'pos')

plt.figure(figsize=(16,4))
# Plot the graph with node positions
nx.draw(G, pos=pos, with_labels=True)
plt.show()
````

### Appendix: Graph drawing

Graph drawing is a field in mathematics and computer science that focuses on the visualization of graphs. A graph is a mathematical structure consisting of nodes (also called vertices) and edges that connect pairs of nodes. Graph drawing aims to represent these nodes and edges visually in order to gain insights into the structure and relationships of the graph.

In graph drawing, various techniques and algorithms are used to create visual representations of graphs. The goal is to create clear and informative visualizations that convey the important properties and patterns of the graph. This can involve positioning the nodes in a two-dimensional or three-dimensional space, determining the layout of the edges, and applying aesthetic criteria to enhance the readability and understandability of the graph.

Graph drawing has applications in various fields, including network analysis, social network analysis, data visualization, computer algorithms, and software engineering. It plays a crucial role in understanding complex systems, identifying patterns and clusters, and facilitating analysis and decision-making processes.

### Some algorithms for graph drawing

1. Force-directed algorithms: Force-directed algorithms aim to position the nodes of a graph in such a way that the edges are evenly distributed and the graph has a visually pleasing layout. These algorithms simulate physical forces between the nodes, such as attractive forces between connected nodes and repulsive forces between all nodes. The nodes are then moved iteratively based on these forces until an equilibrium state is reached.

2. Spring Embedding: Spring embedding is a type of force-directed algorithm where the nodes are treated as particles connected by springs. The attractive force between connected nodes is modeled using Hooke's law, and the repulsive force between all nodes is modeled using Coulomb's law. The nodes are then moved based on the net forces acting on them until the system reaches equilibrium.

3. Fruchterman-Reingold Algorithm: The Fruchterman-Reingold algorithm is another force-directed algorithm that aims to create a visually balanced layout of a graph. It uses a similar approach as spring embedding, but with a different force model. The attractive force between connected nodes is inversely proportional to the distance between them, while the repulsive force between all nodes is proportional to their proximity. The nodes are then moved iteratively based on these forces until the system reaches equilibrium.

4. Layered Drawing Algorithms: Layered drawing algorithms are commonly used for drawing directed acyclic graphs (DAGs) or hierarchical graphs. These algorithms organize the nodes into layers based on their dependencies or hierarchy and then position the nodes within each layer. The edges are typically drawn as straight lines or curves connecting the nodes in adjacent layers.

5. Circular Layout: Circular layout algorithms arrange the nodes of a graph in a circular pattern. The nodes are placed on a circle, and the edges are drawn as arcs connecting the nodes. This layout is often used for visualizing cyclic graphs or graphs with a cyclic component.

6. Spectral Layout: Spectral layout algorithms use the eigenvectors of the graph's adjacency matrix or Laplacian matrix to determine the positions of the nodes. These algorithms leverage the spectral properties of the matrix to embed the graph in a low-dimensional space. The nodes are then placed based on the coordinates derived from the eigenvectors.

These are just a few examples of graph drawing algorithms, and there are many more variations and techniques available. The choice of algorithm depends on the specific requirements of the graph and the desired visual representation.
